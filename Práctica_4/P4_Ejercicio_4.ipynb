# Problema 4
Un conjunto cilindro-pistón contiene 20 Kg de agua, de los cuales 12 están en forma líquida y el resto como vapor. El pistón ejerce una presión absoluta constante de 8 bar. Se entrega energía térmica lentamente hasta que la temperatura del agua asciende a los 220 °C.

**Determinar:**    
a) Título inicial y temperatura inicial.    
b) Trabajo intercambiado con el medio exterior.    
c) Calor intercambiado con el medio exterior.    
d) Representar en un diagrama P-v.   

```python
!pip install CoolProp; # (instala biblioteca CoolProp en CoLab)
import CoolProp.CoolProp as cp;
```

## Solución

Cálculo de x inicial
$$x_i = 1 - \frac{12}{20}$$

$$x_i = 0,4$$

Temperatura inicial
$$t_i = 170,41^\circ \mathrm{C}$$

Volumen específico inicial
$$v = v' + x (v'' - v')$$
$v' = 0,001115 \; \mathrm{m^3/kg}$    
$v'' = 0,2403 \; \mathrm{m^3/kg}$

$$v_i = 0,001115 + 0,4(0,2403 - 0,001115)$$

$$v_i = 0,096789 \; \mathrm{m^3/kg}$$

Volumen específico final   
$v_f = 0,27402 \; \mathrm{m^3/kg}$

$$W_{\Delta V} = Pm (v_f - v_i)$$

$$W_{\Delta V} = \mathrm{(800 \; kPa)(20 \; kg)(0,27402 \; m^3/kg - 0,096789 \; m^3/kg)}$$

$$W_{\Delta V} = 2835,7 \; \mathrm{kJ}$$

Calor intercambiado con el medio exterior

$$u_i = u' + x_i(u'' - u')$$

$u' = 720,05 \; \mathrm{kJ/kg}$    
$u'' = 2575,25 \; \mathrm{kJ/kg}$

$$u_i = 720,05 + 0,4(2575,25 - 720,05)$$

$$u_i = 1462,13 \; \mathrm{kJ/kg}$$

$$u_f = u'' = 2575,25 \; \mathrm{kJ/kg}$$

$$\Delta U = m(u_f - u_i)$$

$$\Delta U = \mathrm{(20 \; kg)(2575,25 \; kJ/kg - 1462,13 \; kJ/kg)}$$

$$\Delta U = 22262,4 \; \mathrm{kJ}$$

Entonces:
$$Q = \Delta U + W_{\Delta V}$$

$$Q = \mathrm{22262,4 \; kJ + 2835,7 \; kJ}$$

$$Q = 25098,1 \; \mathrm{kJ}$$

```python
# Datos:
m = 20 # [kg]
x = 1 - (12/20)
P = 8e5 # [Pa]
T_f = 220 + 273.15 # [K]
```

```python
T_i = cp.PropsSI('T','P',P,'Q',x,'Water')
```

```python
#@title a) Título inicial y temperatura inicial
print("x_i = %.2f"%x)
print("T_i = %.2f [K]"%T_i)
```

```python
# A temperatura final
cp.PhaseSI('P',P,'T',T_f,'Water')
```

```python
# Volumen específico inicial
v_i = 1/cp.PropsSI('D','P',P,'Q',x,'Water')
# Volumen específico final
v_f = 1/cp.PropsSI('D','P',P,'T',T_f,'Water')
# Volumen inicial
V_i = m*v_i
# Volumen final
V_f = m*v_f
# Variación de volumen
ΔV = V_f - V_i
# Trabajo por variación de volumen
W_ΔV = P*ΔV
```

```python

#@title b) Trabajo intercambiado con el medio exterior
print("W_ΔV = %.2f [kJ]"%(W_ΔV/1000))
```

$$Q = \Delta U + W_{\Delta V}$$

```python
# Energía interna específica inicial
u_i = cp.PropsSI('U','P',P,'Q',x,'Water')
# Energía interna específica final vapor saturado
u_f = cp.PropsSI('U','P',P,'Q',1,'Water')
# Energía interna inicial
U_i = m*u_i
# Energía interna final
U_f = m*u_f
# Variación de energía interna
ΔU = U_f - U_i
# Calor intercambiado con el medio
Q = ΔU + W_ΔV
```

```python
#@title c) Calor intercambiado con el medio
print("Q = %.2f [kJ]"%(Q/1000))
```
