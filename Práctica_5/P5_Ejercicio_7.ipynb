### 2) Trabajo entregado por la turbina

$$-W = \Delta H$$

$$-W = \dot{m}_2 (h_3 - h_2)$$

$$W = \dot{m}_2 (h_2 - h_3)$$

```python
h_3 = 2853.46e3 # [J/kg]    (con P=1.4 bar y t=190¬∞C)
```

Intercambiador de calor
$$m_3 h_3 - m_4 h_4 = m_6 h_6 - m_5 h_5$$

$m_3 = m_4$    
$m_5 = m_6$
$$m_3(h_3 - h_4) = m_5 (h_6 - h_5)$$

$$m_3 = \frac{m_5 (h_6 - h_5)}{h_3 - h_4}$$

```python
h_4 = 2733.66e3 # [J/kg]   (con P=1.3 bar y t=130¬∞C)
h_5 =   83.95e3 # [J/kg]   (con P=1 bar y t=20¬∞C)
h_6 =  251.16e3 # [J/kg]   (con P=1 bar y t=60¬∞C)
m_5 =  8 # [kg/s]
```

```python
# Calculo de caudal m√°sico de vapor
m_3 = m_5*(h_6 - h_5)/(h_3 - h_4)
```
