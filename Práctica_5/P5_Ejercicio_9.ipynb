# Ejercicio 9
Despreciando la variación de $\Delta Ec$ y $\Delta Ep$, para el siguiente proceso. Determinar:
1. Caudal másico de vapor $m_1$
2. Caudal másico de oxígeno $m_4$

image.png

## Solución

```python
# Datos:
P_1 = 10e5 # [Pa]
T_1 = 400 + 273.15 # [K]
W_T = 6000e3 # [W]
P_2 = 1e5 # [Pa]
T_2 = 140 + 273.15 # [K]
P_3 = 0.95e5 # [Pa]
x_3 = 1
T_4 = 27 + 273.15 # [K]
T_5 = 150 + 273.15 # [K]
cp = 0.9185e3 # [J/(kg·K)]  (oxígeno)
```

### 1) Caudal másico de vapor $\dot{m}_1$

```python
h_1 = 3264.40e3 # [J/kg]     (P=10 bar y t=400°C)
h_2 = 2756.43e3 # [J/kg]     (P=1 bar y t=140°C)
```

Turbina
$$Q - W = \Delta H$$
$Q = 0$

$$-W = m_1(h_2 - h_1)$$

$$W = m_1 (h_1 - h_2)$$

Despejo $m_1$
$$m_1 = \frac{W}{h_1 - h_2}$$

```python
# Cálculo de m_1
m_1 = W_T/(h_1 - h_2)
```

```python
#@title Resultado
print("m_1 = %.2f [kg/s]"%m_1)
```

---
### 2) Caudal másico de oxígeno $\dot{m}_4$

Intercambiador de calor
$$m_2 h_2 - m_3 h_3 = m_5 h_5 - m_4 h_4$$

$m_2 = m_3$   
$m_4 = m_5$

$$m_2 (h_2 - h_3) = m_4 (h_5 - h_4)$$

$$m_2 (h_2 - h_3) = m_4 c_p (T_5 - T_4)$$

$$m_4 = \frac{m_2 (h_2 - h_3)}{c_p (T_5 - T_4)}$$

```python
m_2 = m_1
h_3 = 2673.2e3 # [J/kg]    (P=0.95 bar y x=1)
```

```python
# Cálculo de m_4
m_4 = m_2*(h_2 - h_3)/(cp*(T_5 - T_4))
```

```python
#@title Resultado
print("m_4 = %.2f [kg/s]"%m_4)
```
