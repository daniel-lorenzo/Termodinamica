### 1) Estado de vapor en 2

```python
# C√°mara de mezcla
P_2 = P_4 = P_3
```

Entalp√≠a en 1
$$h_1 = x_1 (h''_1 - h'_1) + h'_1$$

```python
# De tabla (P=40 bar)
hp_1  =  1087.4e3 # [J/kg]
hpp_1 = 2800.34e3 # [J/kh]
```

```python
h_1 = x_1*(hpp_1 - hp_1) + hp_1
```

```python
# V√°lvula
h_2 = h_1
m_2 = m_1
```

```python
print("h_2 = %.2f [kJ/kg]"%(h_2/1000))
```

$$x_2 = \frac{h_2 - h'_2}{h''_2 - h'_2}$$

```python
# De tabla (P=10 bar)
hp_2  =  762.61e3 # [J/kg]
hpp_2 = 2776.17e3 # [J/kg]
T_2 = 179.88 + 273.15 # [K]
```

```python
x_2 = (h_2 - hp_2)/(hpp_2 - hp_2)
```

```python
#@title Resultados
print("P_2 = %.1f [kPa]"%(P_2/1000))
print("T_2 = %.2f [K]"%T_2)
print("t_2 = %.2f¬∞C"%(T_2-273.15))
print("x_2 = %.3f"%x_2)
```

### 2) Caudal m√°sico en 3 y su estado de vapor

```python
# De tabla
h_4 = 3264.4e3 # [J/kg]
```

$$h_5 = x_5 (h''_5 - h'_5) + h'_5$$

```python
# De tabla
hp_5  =  137.77e3 # [J/kg]
hpp_5 = 2561.59e3 # [J/kg]
```

```python
h_5 = x_5*(hpp_5 - hp_5) + hp_5
```

Turbina   
$Q=0$
$$Q - W = \Delta H$$

$$-W = m_4 (h_5 - h_4)$$

$$W = m_4 (h_4 - h_5)$$

Despejo $m_4$
$$m_4 = \frac{W}{h_4 -h_5}$$

```python
# C√°lculo de m_4
m_4 = W_T/(h_4 - h_5)
```

Caudal m√°sico en 3
$$m_2 + m_3 = m_4$$

$$m_3 = m_4 - m_2$$

```python
# C√°lculo de m_3
m_3 = m_4 - m_2
```

```python
print("m_3 = %.3f [kg/s]"%m_3)
```

$$m_2 h_2 + m_3 h_3 = m_4 h_4$$

Despejo $h_3$
$$h_3 = \frac{m_4 h_4 - m_2 h_2}{m_3}$$

```python
# C√°lculo de h_3
h_3 = (m_4*h_4 - m_2*h_2)/m_3
```

```python
print("h_3 = %.2f [kJ/kg]"%(h_3/1000))
```
