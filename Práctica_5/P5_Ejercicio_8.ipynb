# Ejercicio 8
Para el siguiente proceso determinar:

a) Estado del vapor en el punto 4 (P, t y X)   
b) W recibido por el compresor

## Solución

```python
# Datos:
m_1 = 1.2 # [kg/s]
P_1 = 1e5 # [Pa]
T_1 = 170 + 273.15 # [K]
T_2 = 330 + 273.15 # [K]
m_3 = 0.6 # [kg/s]
P_3 = 3e5 # [Pa]
x_3 = 0.8
m_2 = m_1
```

```python
# Cámara de mezcla
P_2 = P_4 = P_3
m_4 = m_2 + m_3
```

```python
# De tabla
h_1 = 2815.98e3 # [J/kg]     (P=1 bar y t=170°C)
h_2 = 3130.89e3 # [J/kg]     (P=3 bar y t=330°C)
```

$$h_3 = x_3(h''_3 - h'_3) + h'_3$$

```python
hp_3  =  561.43e3 # [J/kg]    (líquido saturado P=3 bar)
hpp_3 = 2724.66e3 # [J/kg]    (vapor saturado P=3 bar)
```

```python
h_3 = x_3*(hpp_3 - hp_3) + hp_3
```

Cámara de mezcla
$$m_2 h_2 + m_3 h_3 = m_4 h_4$$

$$h_4 = \frac{m_2 h_2 + m_3 h_3}{m_4}$$

```python
h_4 = (m_2*h_2 + m_3*h_3)/m_4
```

```python
#@title Resultados
print("P_4 = %.2f [kPa]"%(P_4/1000))
print("h_4 = %.2f [kJ/kg]"%(h_4/1000))
print("t_4 = 193.1°C")
print("No existe X")
```

$$-W = \Delta H$$

$$-W = m_1(h_2 - h_1)$$

$$W = m_1(h_1 - h_2)$$

```python
# Cálculo trabajo recibido por el compresor
Wc = m_1*(h_1 - h_2)
```

```python
#@title Resultado
print("Wc = %.2f [kW]"%(Wc/1000))
```
