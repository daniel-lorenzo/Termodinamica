Para el siguiente sistema y despreciando la variación de energía cinética y potencial, determinar:

a) $\Delta S_u$, (variación de entropía del universo)   
b) $W_{perd}$ (trabajo perdido)    
c) $\eta_{ex}$ (rendimiento exergético) del proceso.

## Solución

```python
# Datos:
T_f1 = 800 # [K]
T_f2 = 310 # [K]
Q_1 = 700 # [kJ/s]
T_0 = 300 # [K]
P_0 = P_1 = 100 # [kPa]
T_1 = 170+273 # [K]
m_3 = 0.6 # [kg/s]
P_3 = 300 # [kPa]
x_3 = 0.8
η_iso = 0.75 # (rend. iso del comp)
η_ex = 0.8   # (rend. ex de la MT)
```

### Análisis MT (máquina térmica)

Rendimiento exergético de la máquina térmica
$$\eta_{exMT} = \frac{W_{MT} + |Q_{u2}|}{|Q_{u1}|}$$

Trabajo de la máquina térmica
$$W_{MT} = |Q_1| - |Q_2|$$

Calor útil de la fuente 1
$$Q_{u1} = |Q_1| - T_0 \Delta S_{F1}$$

Calor útil de la fuente 2
$$Q_{u2} = |Q_2| - T_0 \Delta S_{F2}$$

Variación de entropía de la fuente 1
$$\Delta S_1 = \frac{|Q_1|}{T_1}$$

Variación de entropía de la fuente 2
$$\Delta S_{F2} = \frac{|Q_2|}{T_2}$$

Reemplazando:
$$\eta_{exMT} = \frac{Q_1 - |Q_2| + |Q_2| - T_0 |Q_2|/T_2}{|Q_{u1}|}$$

Despejando $Q_2$
$$|Q_2| = \frac{|Q_1 - \eta_{exMT} |Q_{u1}| T_2|}{T_0}$$

```python
# Variación de entropía de la fuente 1
ΔS_F1 = Q_1/T_f1
# Calor útil de la fuente 1
Q_u1 = Q_1 - T_0*ΔS_F1
# Calor de la fuente 2
Q_2 = abs(Q_1 - η_ex*Q_u1*T_f2)/T_0
# Variación de entropía de la fuente 2
ΔS_F2 = Q_2/T_f2
# Calor útil de la fuente 1
Q_u2 = Q_2 - T_0*ΔS_F2
# Trabajo de la máquina térmica
W_MT = Q_1 - Q_2
```

```python
#@title Resultados
print("|ΔS_F1| = %.3f [kW/K]"%ΔS_F1)
print("|Q_u1| = %.2f [kW]"%Q_u1)
print("|Q_2| = %.2f [kW]"%Q_2)
print("|ΔS_F2| = %.3f [kW/K]"%ΔS_F2)
print("|Q_u2| = %.2f [kW]"%Q_u2)
print("\n|W_MT| = %.2f [kW]"%W_MT)
```

### Análisis del compresor

$$W_c=W_{MT}$$

Datos:   
$P_1 = 1 \; bar$   
$T_1 = 170^\circ \mathrm{C}$
$$h_1, \; s_1$$

Entropía en 2 reversible
$$s_{2rev} = s_1$$

Datos:    
$P_2 = 3 \; \mathrm{bar}$   
$s_{2rev}$     

Entalpía en 2 reversible (de tabla)
$$h_{2rev}$$

Salto de entalpía ideal en el compresor
$$w_{rev} = h_1 - h_{2rev}$$

Salto de entalpía real en el compresor
$$w = \frac{w_{rev}}{\eta_{iso}}$$

Entalpía en 2
$$h_2 = h_1 - w$$

Flujo másico 1
$$m_1 = \frac{W_c}{w}$$

```python
W_c = W_MT
# P_1 = 1 bar, T_1 = 170°C
h_1 = 2815.98 # [kJ/kg]
s_1 = 7.70530 # [kJ/(kg*K)]
s_2rev = s_1
# P_2 = 3 bar
h_2rev = 3069.65 # [kJ/kg]
# Salto de entalpía ideal en el compresor
w_rev = h_1 - h_2rev
# Salto de entalpía real en el compresor
w = w_rev/η_iso
# Entalpía en 2
h_2 = h_1 - w
# Flujo másico 1
m_1 = abs(W_c/w)
```

```python
#@title Resultado
print("m_1 = %.2f [kg]"%m_1)
```

### Cámara de mezcla

En el punto 3:    
$P_3 = 3 \; \mathrm{bar}$   
$x_3 = 0,8$
$$h_3 = x_3 (h''_3 - h'_3) + h'_3$$
$$s_3 = x_3 (s''_3 - s'_3) + s'_3$$

Para la cámara de mezcla tenemos    
$m_2 = m_1$   
$$m_2 + m_3 = m_4$$
$$m_2 h_2 + m_3 h_3 = m_4 h_4$$

Despejando $h_4$
$$h_4 = \frac{m_2 h_2 + m_3 h_3}{m_4}$$

Luego con $P_3$ y $h_4$, obtenemos $s_4$ de tabla.

Variación de entropía entre el punto 1 y el punto 4

$$\Delta S_{14} = m_1 (s_4 - s_1)$$

Variación de entropía entre el punto 3 y el punto 4

$$\Delta S_{34} = m_3 (s_4 - s_3)$$

Variación de entropía del universo
$$\Delta S_u = \Delta S_{14} + \Delta S_{34} + (-\Delta S_{F1}) + \Delta S_{F2}$$

Trabajo perdido
$$W_{perd} = T_0 \Delta S_u$$

Variación de exergía entre 1 y 4
$$\Delta Ex_{14} = m_1 (h_4 - h_1) - T_0 \Delta S_{14}$$

Variación de exergía entre 3 y 4
$$\Delta Ex_{34} = m_3 (h_4 - h_3) -T_0 \Delta S_{34}$$

Rendimiento exergético
$$\eta_{ex} = \frac{\Delta Ex_{14} + \Delta Ex_{34} + Q_{u2}}{|Q_{u1}|}$$

```python
# P_3 = 3 bar
hp_3 =   561.43 # [kJ/kg]
hpp_3 = 2724.66 # [kJ/kg]
sp_3 =  1.67164 # [kJ/(kg·K)]
spp_3 = 6.99090 # [kJ/(kg·K)]

h_3 = x_3*(hpp_3 - hp_3) + hp_3
s_3 = x_3*(spp_3 - sp_3) + sp_3

m_2 = m_1
m_4 = m_2 + m_3

h_4 = (m_2*h_2 + m_3*h_3)/m_4
```

```python
#@title Resultados
print("m_4 = %.3f [kg/s]"%m_4)
print("h_4 = %.2f [kJ/kg]"%h_4)
```

```python
# P_3 = 3 bar y h_4 = 2832
s_4 = 7.2222 # [kJ/(kg·K)]
```

```python
ΔS_14 = m_1*(s_4 - s_1)
ΔS_34 = m_3*(s_4 - s_3)
ΔS_u = ΔS_14 + ΔS_34 + (-ΔS_F1) + ΔS_F2
W_perd = ΔS_u*T_0
ΔEx_14 = m_1*(h_4 - h_1) - T_0*ΔS_14
ΔEx_34 = m_3*(h_4 - h_3) - T_0*ΔS_34
η_ex = (ΔEx_14 + ΔEx_34 + Q_u2)/abs(Q_u1)
```

```python
#@title Resultados
print("ΔS_14 = %.4f [kJ/K]"%ΔS_14)
print("ΔS_34 = %.4f [kJ/K]"%ΔS_34)
print("ΔS_u = %.4f [kJ/kg]"%ΔS_u)
print("W_perd = %.2f [kW]"%W_perd)
print("ΔEx_14 = %.2f [kW]"%ΔEx_14)
print("ΔEx_34 = %.2f [kW]"%ΔEx_34)
print("\nη_ex = %.2f"%η_ex)
```
