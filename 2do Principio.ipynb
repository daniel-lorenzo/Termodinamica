## Segundo principio de la termodinámica 

La energía es una propiedad conservada y no se sabe de ningún proceso que viole la primera ley de la termodinámica. Sin embargo, satisfacerla no asegura que en realidad el proceso tenga lugar.
La segunda ley de la termodinámica establece que un proceso ocurre en cierta dirección y no en la dirección contraria. Una tasa de café caliente no se pondrá más calienteen una habitación fría. Transferir calor a un alambre no generará electricidad. Transferir calor a una rueda de paletas no hará que ésta gire.

Sin embargo, el uso de la segunda ley de la termodinámica no se limita a identificar la dirección de los procesos, también afirma que la energía tiene calidad así como cantidad. La segunda ley de la termodinámica se usa también para determinar los límites teóricos en el desempeño de sistemas de ingeniería de uso ordinario, como máquinas térmicas y refrigeradores, así como predecir el grado de terminación de las reacciones químicas.
Un proceso no ocurre a menos que satisfaga tanto la primera como la segunda leyes de la termodinámica.

Es muy conveniente tener un hipotético cuerpo que posea una capacidad de energía térmica relativamente grande (masa x calor específico) que pueda suministrar o absorber cantidades finitas de calor sin experimentar ningún cambio de temperatura. Tal cuerpo se llama depósito de energía térmica, o sólo depósito. En la práctica, los grandes cuerpos de agua, como océamos, lagos y ríos, así como el aire atmosférico, se pueden modelar como depósitos de energía térmica debido a sus grandes capacidades de almacenaje de energía o masas térmicas.

Un depósito que suministra energía en la forma de calor se llama fuente, y otro que absorbe energía en la forma de calor se llama sumidero.

El trabajo se puede convertir en calor de manera directa y por completo, pero convertir el calor en trabajo requiere algunos dispositivos especiales. Estos dispositivos se llaman máquinas térmicas. 
Se caracterizan por:
1. Reciben calor de una fuente a alta temperatura (energía solar, horno de petróleo, reactor nuclear, etc.)
2. Convierten parte de este calor en trabajo (por lo general en la forma de una flecha rotatoria)
3. Rechazan el calor de desecho hacia un sumidero de calor de baja temperatura (la atmósfera, los ríos ,etc)
4. Operan en un ciclo

El dispositivo productor de trabajo que mejor se ajusta a la definición de una máquina térmica es la central eléctrica de vapor, la cual es una máquina de combustión externa.

La eficiencia de una máquina térmica se define como

$$ \eta_\mathrm{ter} = \frac{W_\mathrm{neto,salida}}{Q_\mathrm{entrada}} = \frac{Q_\mathrm{entrada}-|Q_\mathrm{salida}|}{Q_\mathrm{entrada}} = 1 - \frac{Q_\mathrm{salida}}{Q_\mathrm{entrada}} $$

donde $W_\mathrm{neto,salida}$ es la salida de trabjo neto de la máquina térmica, $Q_H$ la cantidad de calor suministrada a la máquina y $Q_L$ la cantidad de calor que la máquina cede.
La eficiencia térmica de una máquina térmica siempre es menor a la unidad porque $Q_L$ y $Q_H$ se definen como cantidades positivas. Así, incluso con las máquinas térmicas más eficientes disponibles en la actualidad, casi la mitad de la energía suministrada termina en ríos, lagos o en la atmósfera como energía de desecho o inútil.

Los refrigeradores y las bombas de calor son dispositivos que absorben calor de medios de baja temperatura y lo ceden hacia la atmósfera a medios de mayor temperatura. El desampeño de un refrigerador o bomba de calor se expresa en términos de coeficiente de desempeño, definido como

$$ COP_R = \frac{Q_L}{W_\mathrm{neto,entrada}} = \frac{1}{Q_H/Q_L - 1} $$

$$ COP_{HP} = \frac{Q_H}{W_\mathrm{neto,entrada}} = \frac{1}{1 - Q_L/Q_H} $$

El enunciado de __Kelvin-Plank__ de la segunda ley de la termodinámica establece que ninguna máquina térmica pude producir una cantidad neta de trabajo mientras intercambia calor con un solo depósito. Es decir, ninguna máquina térmica puede convertir todo el calor que recibe en trabajo útil.

El Enunciado de __Clausius__ de la segunda ley expresa que ningún depósito puede transferir calor de un cuerpo más frío a otro más caliente sin dejar un efecto sobre los alrededores. El enunciado establece simplemente que un refiregerador no puede operar a menos que su compresor sea propulsado mediante una fuente de energía externa, como un motor eléctrico.

Cualquier dispositivo que viola la primera o segunda ley de la termodinámica se llama máquina de movimiento perperuo (no exixte).

La segunda ley de la termodinámica establece que ninguna máquina térmica puede tener una eficiencia de 100 por ciento. 

Un proceso reversible se define como un proceso que se puede invertir sin dejar ningún rastro en los alrededores. Los procesos reversibles en realidad no ocurren en la naturaleza, sólo son idealizaciones de procesos reales.
Entonces, quizá se pregunte por qué preocuparse de esta clase de procesos ficticios. Hay dos razones: una es que son fáciles de analizar, puesto que un sistema pasa por una serie de estados de equilibrio durante un proceso reversible; y otra es que sirven como modelos idealizados con los que es posible comparar los procesos reales.
Los procesos reversibles pueden ser considerados como límites teóricos para los irreversibles correspondientes. Los efectos que hacen que un proceso sea irreversible son la fricción, la expansión o compresión de no cuasiequilibrio y la transderencia de calor debida a una diferencia finita de temperatura, las cuales se denominan irreversibilidades. El concepto de reversible conduce a la definición de eficiencia según la segunda ley para procesos reales. Mientras mejor sea el diseño, menores son la irreversibilidades y mayor la eficiencia según la segunda ley.

El __ciclo de Carnot__ opera entre dos fuentes de calor a temperaturas $T_H$ y $T_L$ y usa sólo procesos reversibles. Su eficiencia térmica depende únicamente de $T_H$ y $T_L$. Otro planteamiento equivalente de la segunda ley es que ninguna máquina que opere entre dos temperaturas dadas puede ser más eficiente que una máquina de Carnot que opera entre las mismas temperaturas.

Es un ciclo reversible compuesto por cuatro procesos reversibles, dos isotérmicos y dos adiabáticos. Los principios de Carnot establecen que las eficiencias térmicas de las máquinas térmicas reversibles que operan entre dos depósitos son las mismas, y que ninguna máquina de este tipo es más eficiente que una reversible que opera entre los mismos dos depósitos. Estos enunciados crean el fundamento para establecer una escala termodnámica de temperatura relacionada con las transderencias de calor entre un dispositivo reversible y los depósitos de alta y baja temperaturas, por medio de

$$ \left( \frac{Q_H}{Q_L} \right)_\mathrm{rev} = \frac{T_H}{T_L} $$

Por lo tanto, la relación de $Q_H/Q_L$ se puede reemplazar por $T_H/T_L$ para dispositivos reversibles, donde $T_H$ y $T_L$ son las temperaturas absolutas de los depósitos de temperaturas alta y baja, repesctivamente.

Una máquina térmica que opera en un ciclo reversible de carnot se llama máuina térmica de Carnot. La efifiencia térmica de una máquina térmica de Carnot, así como de las otras máquinas térmicas reversibles está expresada por

$$ \eta_{ter,rev} = 1 - \frac{T_L}{T_H} $$

Esta es la eficiencia máxima que puede tener una máquina térmica que opera entre dos depósitos a temperaturas $T_H$ y $T_L$.

Los COP de refrigeradores reversibles y las bombas de calor se obtienen de una manera similar

$$ COP_\mathrm{R,rev} = \frac{1}{T_H/T_L -1} \quad ; \quad COP_\mathrm{HP,rev} = \frac{1}{1 - T_L/T_H}  $$

De nuevo, éstos son los COP más altos que puede tener un refrigerador o una bomba de calor que opera entre los límites de temperatura $T_H$ y $T_L$

## Entropía
La segunda ley de la termodinámica conduce a la definición de una nueva propiedad llamada entropía que es una medida cuantitativa de desorden microscópico para un sistema. Un planteamiento importante de la segunda ley de la termodinámica es que la entropía de un sistema aislado puede aumentar pero nunca disminuir. Si un sistema interactúa con su entorno, el cambio total de entropía del sistema y el entorno nunca puede ser negativo. Si la interacción implica sólo procesos reversibles, la entropía total es constante y $\Delta S = 0$; si hay procesos irreversibles, la entropía total aumenta y $\Delta S > 0$.
Cualquier cantidad cuya integral cíclica es cero es un propiedad, y la entropía está definida como

$$ dS = \left( \frac{dQ}{T} \right)_\mathrm{int,rev} $$

Para el caso especial de un proceso isotérmico internamente reversible,

$$ \Delta S = \frac{Q}{T_0} $$

La parte de la desigualdad en la desigualdad de Clausius combinada con la definición de entropía produce una desigualdad conocida como el principio de incremento de entropía, que se expresa como

$$ S_\mathrm{gen} \geq 0 $$

donde $S_\mathrm{gen}$ es la entropía generada durante el proceso. El cambio de entropía es ocasionado por la transferencia de calor, el flujo másico e irreversibilidades. La transferencia de calor desde un sistema la disminuye. El efecto de las irreversibilidades siempre es aumentar la entropía.

El cambio de entropía y las reacciones isentrópicas para un proceso pueden resumirse como:
1. Sustancias puras:

Cualquier proceso:
$$ \Delta s = s_2 - s_1 $$
Proceso isentrópico:
$$ s_2 = s_1 $$
2. Sustancias incompresibles:

Cualquier proceso:
$$ s_2 - s_1 = c_\mathrm{prom} \ln \frac{T_2}{T_1} $$
Proceso isentrópico:
$$ T_2 = T_1 $$
3. Gases ideales:  
a. Calores específicos constantes (tratamiento aproximado):
$$ \begin{array}{l}
\displaystyle s_2 -s_1 = c_{v,\mathrm{prom}} \ln \frac{T_2}{T_1} + R \ln \frac{V_2}{V_1} \\
\displaystyle s_2 -s_1 = c_{p\mathrm{,prom}} \ln \frac{T_2}{T_1} + R \ln \frac{P_2}{P_1} 
\end{array}
$$

Procesos isentrópicos:
$$ \begin{array}{l}
\displaystyle \left( \frac{T_2}{T_1} \right)_{s=\mathrm{const}} = \left( \frac{V_1}{V_2} \right)^{k-1}  \\
\displaystyle \left( \frac{T_2}{T_1} \right)_{s=\mathrm{const}} = \left( \frac{P_2}{P_1} \right)^{(k-1)/k} \\
\displaystyle \left( \frac{P_2}{P_1} \right)_{s=\mathrm{const}} = \left( \frac{V_1}{V_2} \right)^k 
 \end{array}
 $$
 b. Calores específicos variables (tratamiento exacto):  
 Cualquier proceso:
 $$ s_2 - s_1 = s_2^\circ - s_1^\circ - R \ln \frac{P_2}{P_1} $$
 Proceso isentrópico:
 $$ s_2^\circ = s_1^\circ + R \ln \frac{P_2}{P_1} $$
 
 $$\begin{array}{l}
 \displaystyle \left( \frac{P_2}{P_1} \right)_{s=\mathrm{const}} = \frac{P_{r2}}{P_{r1}} \\
 \displaystyle \left( \frac{V_2}{V_1} \right)_{s=\mathrm{const}} = \frac{V_{r2}}{V_{r1}}
 \end{array}
 $$
 donde $P_r$ es la presión relativa y $V_r$ es el volumen específico relativo. La función $s^\circ$ sólo depende de la temperatura.
 
 El trabajo de flujo estacionario para un proceso reversible puede expresarse en términos de las propiedades del fluido como
 
 $$ w_\mathrm{rev} = - \int_1^2 V \; dP - \Delta ec - \Delta ep $$
 
 Para sustancias incompresibles $(V = \mbox{constante})$ se simplifica a
 
 $$ w_\mathrm{rev} = -V (P_2 - P_1) - \Delta ec - \Delta ep $$
 
 El trabajo realizado durante un proceso de flujo estacionario es proporcional al volumen específico, por consiguiente $V$ debe mantenerse tan pequeño como sea posible durante un proceso de compresión para minimizar la entrada de trabajo y tan grande como sea posible durante un proceso de expansión para aumentar al máximo la salida de trabajo.
 
Las entradas de trabajo reversibles de un compresor que comprime un gas ideal de $T_1$, $P_1$ a $P_2$ de manera isentrópica $(P V^k = \mbox{constante})$, se determina mediante la integración para cada caso, de donde se obtienen los siguientes resultados:

$$ \begin{array}{l}
\displaystyle \mbox{Isentrópica: } w_\mathrm{comp,ent} = \frac{kR(T_2 - T_1)}{k -1} = \frac{kRT_1}{k - 1} \left[ \left( \frac{P_2}{P_1} \right)^{(k-1)/k} - 1 \right] \\
\displaystyle \mbox{Politrópica: } w_\mathrm{comp,ent} = \frac{nR(T_2 - T_1)}{n -1} = \frac{nRT_1}{n - 1} \left[ \left( \frac{P_2}{P_1} \right)^{(n-1)/n} - 1 \right] \\
\displaystyle \mbox{Isotérmica: } w_\mathrm{comp,ent} = RT \ln \frac{P_2}{P_1}
\end{array}
$$
Es posible reducir la entrada de trabajo a un compresor usando la compresión de etapas múltiples con interenfriamiento. Para lograr ahorros máximos en la entrada de trabajo, las razones de presión por cada etapa del compresor deben ser iguales.

La mayoría de los dispositivos de flujo estacionario operan bajo condiciones adiabáticas, y el proceso ideal para estos dispositivos es el isentrópico. El parámetro que describe qué tanta eficiencia tiene un dispositivo para acercarse al dispositivo isentrópico correspondiente se llama eficiencia isentrópica o adiabática, se expresa como sigue:
$$ \eta_T = \frac{\mbox{Trabajo real de la turbina}}{\mbox{Trabajo isentrópico de la turbina}} = \frac{w_a}{w_s} \approx \frac{h_1 - h_{2a}}{h_1 - h_{2s}}$$

$$ \eta_C = \frac{\mbox{Trabajo isentrópico del compresor}}{\mbox{Trabajo real del compresor}} = \frac{w_s}{w_a} \approx \frac{h_{2s} - h_1}{h_{2a} - h_1}$$

$$ \eta_N = \frac{\mathrm{Energía \; cinética \; real \; a \\ \; la \; salida \; de \; la \; tobera}}{\mathrm{Energía \; cinética \; isentrópica \\ a \; la \; salida \; de \; la \; tobera}} = \frac{V^2_{2a}}{V^2_{2s}} \approx \frac{h_1 - h_{2a}}{h_1 - h_{2s}} $$

En las relaciones anteriores, $h_{2a}$ y $h_{2s}$ son los valores de la entalpía en el estado de salida para los procesos real e isentrópico, respectivamente.

El balance de entropía para cualquier sistema que experimenta cualquier proceso puede expresarse en forma general como

$$ \underbrace{S_\mathrm{ent} - S_\mathrm{sal} }_\mathrm{Tasa \; de \; transferencia \\ neta \; de \; entropía \; por \\ calor \; y \; masa} + \underbrace{ S_\mathrm{gen} }_\mathrm{Generación \\ de \; entropía} = \underbrace{ \Delta S_\mathrm{sistema}}_\mathrm{Cambio \; de \\ entropía} $$

o, en forma de tasa, como

$$ \underbrace{ \dot{S}_\mathrm{ent} - \dot{S}_\mathrm{sal} }_\mathrm{Tasa \; de \; transferencia \\ neta \; de \; entropía \; por \\ calor \; y \; masa} + \underbrace{ \dot{S}_\mathrm{gen} }_\mathrm{Tasa \; de \; genera-\\ción \; de \; entropía} = \underbrace{ d S_\mathrm{sistema} / dt }_\mathrm{Tasa \; de \; cambio \\ de \; entropía} $$

Para un proceso general de flujo estacionario se simplifica a

$$ \dot{S}_\mathrm{gen} = \sum \dot{m}_e s_e - \sum \dot{m}_i s_i - \sum \frac{\dot{Q}_k}{T_k} $$
