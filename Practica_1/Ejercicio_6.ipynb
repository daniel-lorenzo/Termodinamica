# Ejercicio 6
Un turbogenerador consume 1500 Kg/h de vapor y desarrolla una potencia de 300 kW. Determinar el trabajo por Kg de vapor.

## Solución:

$ \dot{m} = \mathrm{ 1500 \, (Kg/h) \cdot (1 \, h) / (3600 \, s)}$

$ \dot{W} = \mathrm{ 300 \, kW }$

```python
m = 1500*1/3600 # kg/s
W = 300 # kW
```

```python
#@title
print('m = %.4f kg/s'%m)
```

$$ w = \mathrm{ \frac{J}{Kg} = \frac{W}{m} = \frac{kJ}{Kg/s} = \frac{kJ}{Kg} } $$

```python
w = W/m # kJ/kg
```

```python
#@title
print('w = %s kJ/kg'%w)
```

# Ejercicio 6.1
Que cantidad de calor será necesaria para producir un trabajo de 0,25 kWh, sabiendo que se pierde 10% con el medio.

image.png

## Solución
$$ \mathrm{ Potencia = \frac{Trabajo}{Tiempo}  \quad  \Rightarrow \quad Trabajo = Potencia \cdot Tiempo } $$

$$ \mathrm{1 \, h = 3600 \, s} $$

$$ \mathrm{0,25 \, kWh \cdot \frac{3600 \, s}{h} = 900 \, kJ } $$

Teniendo en cuenta que se pierde un 10%

    0.9 ________ 900 kJ
    1 __________ x kJ
    
    x = 1000 kJ

$$ Q = \mathrm{ 1000 \, kJ = (1000 \cdot 0,2388) \, kcal = 238,8 \, kcal }$$

```python
Q = (0.25 * 3600)/0.9
```

```python
#@title
print('Q = %.1f kJ = %.1f kcal'%(Q , Q*0.2388) )
```
