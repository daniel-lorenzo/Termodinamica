# Ejercicio 18
Un recipiente rígido y adiabático contiene una masa de 40 kg de vapor que se encuentra
a 80°C. Si el 37,5% de la masa está en fase líquida, y el resto como vapor:

Calcular:

    a) Título, x
    b) P del recipiente
    c) V del recipiente

## Solución

```python
# instala biblioteca termodinámica, (usar solo en colab)
!pip install CoolProp;
# Carga biblioteca
import CoolProp.CoolProp as cp
```

### a) Título $x$

```python
x = 1 - 37.5 / 100
```

```python
#@title resultado:
print('x =',x)
```

### b) $P$ del recipiente

```python
# Dato:
T = 80 + 273.15 # K
# Presión del recipiente
P = cp.PropsSI('P','T',T,'Q',x,'Water')
```

```python
#@title resultado:
print('P = %.2f kPa'%(P/1000))
```

### c) $V$ del recipiente

```python
# Dato:
m = 40 # kg
vp1 = 1/cp.PropsSI('D','T',T,'Q',0,'Water')  # v'
vp2 = 1/cp.PropsSI('D','T',T,'Q',1,'Water')  # v''
# Volumen específico
v = x*(vp2 - vp1) + vp1
```

```python
#@title
print('Volumen específico:') 
print('v = %.3f m^3/kg'%v)
```

```python
# o tambien
v = 1/cp.PropsSI('D','T',T,'Q',x,'Water')
```

```python
#@title 
print('Volumen específico:') 
print('v = %.3f m^3/kg'%v)
```

```python
# Volumen del recipiente
Vol = m*v
```

```python
#@title resultado:
print('Volumen del recipiente:') 
print('V = %.3f m^3'%Vol)
```
