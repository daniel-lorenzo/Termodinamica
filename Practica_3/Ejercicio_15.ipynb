# Ejercicio 15
Utilizando la tabla de propiedades del agua, determinar:

    a) v, u, h y x si P=20 bar y T=400°C 
    b) v, u, h y T si P=10 bar y x=1
    c) v, u, h y P si se tiene líquido saturado a T=295°C
    d) v, u, T y x si P=2 kPa y h=1500 kJ/kg
    e) x, u, h y T si P=5 bar y v=0,4148 m³/kg
    f) v, u, h y P si T=120°C y x=0,7
    g) v, h, T y x si P=8,5 bar y u=900 kJ/kg

```python
!pip install CoolProp;  # (instala biblioteca termodinámica)
```

```python
import CoolProp.CoolProp as cp
```

## Solución

### a) ítem

```python
# Datos:
P = 20e5 # Pa
T = 400 + 273.15 # K
```

```python
# Volumen específico (v)
v = 1 / cp.PropsSI('D','P',P,'T',T,'Water')
# Energía interna (u)
u = cp.PropsSI('U','P',P,'T',T,'Water')
# Entalpía (h)
h = cp.PropsSI('H','P',P,'T',T,'Water')
# Fase
fase = cp.PhaseSI('P',P,'T',T,'Water')
```

```python
#@title a) Resultados:
print('ítem (a)')
print('v = %.5f m^3/kg'%v)
print('u = %.2f kJ/kg'%(u/1000))
print('h = %.2f kJ/kg'%(h/1000))
print(fase)
```

### b) ítem

```python
# Datos:
P = 10e5 # Pa
x = 1
```

```python
# Volumen específico (v)
v = 1 / cp.PropsSI('D','P',P,'Q',x,'Water')
# Energía interna (u)
u = cp.PropsSI('U','P',P,'Q',x,'Water')
# Entalpía (h)
h = cp.PropsSI('H','P',P,'Q',x,'Water')
# Temperatura
T = cp.PropsSI('T','P',P,'Q',x,'Water')
```

```python
#@title b) Resultados:
print('ítem (b)')
print('v = %.5f m^3/kg'%v)
print('u = %.2f kJ/kg'%(u/1000))
print('h = %.2f kJ/kg'%(h/1000))
print('T = %.2f K'%T)
print('t = %.2f°C'%(T - 273.15))
```

### d) ítem

```python
# Datos:
P = 2e3 # Pa
h = 1500e3 # J/kg
```

$$ h = x (h'' - h') + h' $$

$$ x = \frac{h - h'}{h'' - h'} $$

```python
hp1 = cp.PropsSI('H','P',P,'Q',0,'Water')
hp2 = cp.PropsSI('H','P',P,'Q',1,'Water')

x = (h - hp1)/(hp2 - hp1)
```

```python
# Volumen específico
v = 1 / cp.PropsSI('D','P',P,'Q',x,'Water')
# Energía interna
u = cp.PropsSI('U','P',P,'Q',x,'Water')
# Temperatura
T = cp.PropsSI('T','P',P,'Q',x,'Water')
```

```python
#@title d) Resultados:
print('ítem (d)')
print('v = %.3f m^3/kg'%v)
print('u = %.2f kJ/kg'%(u/1000))
print('T = %.2f K'%T)
print('t = %.2f°C'%(T - 273.15))
print('x = %.2f'%x)
```

### g) ítem

```python
# Datos:
P = 8.5e5 # Pa
u = 900e3 # J/kg
```

```python
cp.PhaseSI('P',P,'U',u,'Water')
```

```python
x = cp.PropsSI('Q','P',P,'U',u,'Water')
```

```python
#@title Resultado:
print('x = %.4f'%x)
```

$$ u = x \,(u'' - u') + u' $$

Entonces:
$$ x = \frac{u - u'}{u'' - u'} $$

```python
up1 = cp.PropsSI('U','P',P,'Q',0,'Water')
up2 = cp.PropsSI('U','P',P,'Q',1,'Water')

x = (u - up1)/(up2 - up1)
```

```python
#@title Resultado:
print('x = %.4f'%x)
```

$$\begin{array}{l}
v = x \, (v'' - v') + v' \\
h = x \, (h'' - h') + h'
\end{array}$$

```python
# Volumen específico
v = 1/cp.PropsSI('D','P',P,'U',u,'Water')
# Entalpía
h = cp.PropsSI('H','P',P,'U',u,'Water')
# Temperatura
T = cp.PropsSI('T','P',P,'U',u,'Water')
```

```python
#@title g) Resultados:
print('ítem (g)')
print('v = %.5f m^3/kg'%v)
print('h = %.2f kJ/kg'%(h/1000))
print('T = %.2f K'%T)
print('t = %.2f°C'%(T - 273.15))
```
