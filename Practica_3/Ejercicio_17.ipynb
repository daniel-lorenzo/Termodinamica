# Ejercicio 17
Una masa de agua líquida saturada de 5kg se evapora totalmente a una presión 
constante de 500 kPa.

Calcular:
    
    a) ΔV
    b) W
    c) Q

## Solución

```python
!pip install CoolProp;  # (instala biblioteca termodinámica)
import CoolProp.CoolProp as cp
```

```python
# Datos:
m = 5 # kg
P = 500e3 # Pa
vp1 = 1/cp.PropsSI('D','P',P,'Q',0,'Water')   # v'
vp2 = 1/cp.PropsSI('D','P',P,'Q',1,'Water')   # v''
```

```python
# Variación de volumen
DV = m*(vp2 - vp1)
```

```python
#@title 
print('DV = %.3f m^3'%DV)
```

$$ W_{\Delta V} = P \, \Delta V $$

```python
W_DV = P*DV
```

```python
#@title
print('W_DV = %.2f kJ'%(W_DV/1000))
```

```python
up1 = cp.PropsSI('U','P',P,'Q',0,'Water')  # u'
up2 = cp.PropsSI('U','P',P,'Q',1,'Water')  # u''

DU = m*(up2 - up1)

Q = DU + W_DV
```

```python
#@title 
print('up1 = %.2f kJ/kg'%(up1/1000))
print('up2 = %.2f kJ/kg'%(up2/1000))
print('DU = %.2f kJ'%(DU/1000))
print('Q = %.2f kJ'%(Q/1000))
```
