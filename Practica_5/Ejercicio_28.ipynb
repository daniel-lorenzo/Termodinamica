## Solución

```python
# Datos:
cp_O2 = 0.9185e3 # [J/(kg·K)]
P_1 = 10e5 # [Pa]
P_2 = 1e5 # [Pa]
P_3 = 0.95e5 # [Pa]
T_1 = 400 + 273.15 # [K]
T_2 = 140 + 273.15 # [K]
T_4 = 27 + 273.15 # [K]
T_5 = 150 + 273.15 # [K]
W_T = 6000e3 # [W]
x_3 = 1
```

```python
# instala coolprop library en colab
!pip install CoolProp # (thermophysical property database)
# carga biblioteca
import CoolProp.CoolProp as cp
```

---
### 1) $\dot{m}_1$

```python
# consultamos en la base de datos h_1 y h_2
h_1 = cp.PropsSI('H','P',P_1,'T',T_1,'Water')
h_2 = cp.PropsSI('H','P',P_2,'T',T_2,'Water')
```

```python
#@title 
print('h_1 = %.2f [kJ/kg]'%(h_1/1000))
print('h_2 = %.2f [kJ/kg]'%(h_2/1000))
```

Para la turbina
$$ Q - W = \Delta H $$

$$ W_T = \dot{m}_1 (h_1 - h_2) \tag{1} $$

Despejo $\dot{m}_1$
$$ \dot{m}_1 = \frac{W_T}{h_1 - h_2} $$

```python
m_agua = W_T/(h_1 - h_2)
```

```python
#@title 
print('m_agua = %.3f (kg/s)'%m_agua)
```

---
### 2) $\dot{m}_4 \quad (O_2)$

Para el intercambiador de calor
$$ Q - W = \Delta H $$

$$ \Delta H = 0 $$

$$ H_e = H_s $$

$$ m_{agua} (h_2 - h_3) = m_{gas} \Delta H_{gas} $$

Entonces
$$ m_{agua} (h_2 - h_3) = m_{gas} c_p (T_5 - T_4)  \tag{2} $$

Despejo $m_{gas}$
$$ m_{gas} = m_{agua} \frac{h_2 - h_3}{c_p (T_5 - T_4)} $$

```python
h_3 = cp.PropsSI('H','P',P_3,'Q',x_3,'Water')

m_gas = m_agua*(h_2 - h_3)/(cp_O2*(T_5 - T_4))
```

```python
#@title 
print('m_gas = %.2f (kg/s)'%m_gas)
```
