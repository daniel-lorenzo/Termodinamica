Determinar:
1. $\dot{W}_A$
2. $\dot{W}_B$
3. $Q_{2-3}$

## Solución

```python
# Datos:
P_1 = 6e6 # [Pa]
T_1 = 400 + 273.15 # [K]
m = 13 # [kg/s]
P_2 = 7e5 # [Pa]
x_1 = 1
T_3 = 400 + 273.15 # [K]
P_3 = P_2
P_4 = 5e3 # [Pa]
x_4 = 0.9
```

```python
!pip install CoolProp
import CoolProp.CoolProp as cp
```

---
### 1) Turbina vapor
$$ Q - W = \Delta H $$

$$ 0 - W_A = \dot{m} \, (h_2 - h_1) $$

$$ W_A = \dot{m} \, (h_1 - h_2) \tag{1}$$

```python
h_1 = cp.PropsSI('H','P',P_1,'T',T_1,'Water')
h_2 = cp.PropsSI('H','P',P_2,'Q',x_1,'Water')
```

```python
#@title 
print('h_1 = %.2f [kJ/kg]'%(h_1/1000))
print('h_2 = %.2f [kJ/kg]'%(h_2/1000))
```

```python
W_A = m*(h_1 - h_2)
```

```python
#@title 
print('W_A = %.2f [kW]'%(W_A/1000))
```

---
### 2) Turbina de vapor

$$ Q - W = \Delta H $$

$$ 0 - W_B = \dot{m} \, (h_4 - h_3) $$

$$ W_B = \dot{m} \, (h_3 - h_4) \tag{2} $$

```python
h_3 = cp.PropsSI('H','P',P_3,'T',T_3,'Water')
```

```python
#@title 
print('h_3 = %.2f (kJ/kg)'%(h_3/1000))
```

$$ h_4 = x_4 \, (h'' - h') + h' $$

```python
hp1 = cp.PropsSI('H','P',P_4,'Q',0,'Water')  #  h'
hp2 = cp.PropsSI('H','P',P_4,'Q',1,'Water')  #  h''

h_4 = x_4*(hp2 - hp1) + hp1
```

```python
#@title 
print('h_4 = %.2f [kJ/kg]'%(h_4/1000))
```

```python
# o tambien
h_4 = cp.PropsSI('H','P',P_4,'Q',x_4,'Water')
```

```python
#@title 
print('h_4 = %.2f [kJ/kg]'%(h_4/1000))
```

```python
W_B = m*(h_3 - h_4)
```

```python
#@title 
print('W_B = %.2f [kW]'%(W_B/1000))
```

---
### 3) Para el intercambiador de calor

$$ Q - W = \Delta H $$

$$ Q - 0 = \Delta H $$

$$ Q_{2-3} = \dot{m} \, (h_3 - h_2) \tag{3}$$

```python
Q_23 = m*(h_3 - h_2)
```

```python
#@title 
print('Q_23 = %.2f [kJ/s]'%(Q_23/1000))
```
