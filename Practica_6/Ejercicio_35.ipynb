Se abre la válvula y el recipiente con vacío inicial, se llena hasta alcanzar la presión de línea:

Determinar:
1. Temperatura final del vapor
2. Masa de vapor

## Solución

$$ H_E = U_f $$

$$ m h_e = m u_f $$

$$ h_e = u_f $$

```python
!pip install CoolProp
import CoolProp.CoolProp as cp
```

```python
# Datos:
T_L = 250 + 273.15 # [K]
x_L = 0.8
V = 0.2 # [m^3]
```

```python
# Presión de línea
P_L = cp.PropsSI('P','T',T_L,'Q',x_L,'Water')
```

```python
# Entalpía específica de entrada
h_e = cp.PropsSI('H','T',T_L,'Q',x_L,'Water')
# Energía interna específica final
u_f = h_e
```

$$ x = \frac{u - u'}{u'' - u'} $$

```python
up1 = cp.PropsSI('U','T',T_L,'Q',0,'Water')  # u'
up2 = cp.PropsSI('U','T',T_L,'Q',1,'Water')  # u''
```

```python
x_2 = (u_f - up1)/(up2 - up1)
```

```python
#@title
print('x_2 = %.3f'%x_2)
```

$$ v = x_2 (v'' - v') + v' $$

```python
vp1 = 1 / cp.PropsSI('D','T',T_L,'Q',0,'Water')
vp2 = 1 / cp.PropsSI('D','T',T_L,'Q',1,'Water')
```

```python
v_2 = x_2*(vp2 - vp1) + vp1
dens = 1/v_2
```

```python
# Cálculo de masa
m = dens*V
```

```python
#@title
print('m = %.2f [kg]'%m)
```

```python
T_2 = cp.PropsSI('T','P',P_L,'Q',x_2,'Water')
```

```python
#@title 
print('T_2 = %.2f [K]'%T_2)
print('t_2 = %.2f [°C]'%(T_2 - 273.15))
```
