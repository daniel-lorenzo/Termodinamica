# Ejercicio 6
Para el siguiente proceso determinar:

a) Estado de vapor en 2 y 4     
b) Potencia Térmica Q     
c) Potencia de la Turbina    
d) Variación de Entropía del Universo   
e) Trabajo perdido   


image.png

## Solución

### a) Estado de vapor en 2 y 4

__Punto 1:__   
$m_1, \; P_1, \; x_1 \qquad \rightarrow \qquad T_1, \; h'_1, \; h''_1, \; s'_1, \; s''_1$

$$h_1 = x_1 (h''_1 - h'_1) + h'_1$$

$$s_1 = x_1 (s''_1 - s'_1) + s'_1$$

Para una válvula
$$h_2 = h_1$$

__Punto 2:__   
$P_2, \; h_2 \qquad \rightarrow \qquad h'_2, \; h''_2, \; T_2, \; s'_2, \; s''_2$

$$x_2 = \frac{h_2 - h'_2}{h''_2 - h'_2}$$

$$s_2 = x_2 (s''_2 - s'_2) + s'_2$$

En la cámara de mezcla
$$m_4 = m_2 + m_3$$

$$m_4 h_4 = m_2 h_2 + m_3 h_3$$

$$P_4 = P_3 = P_2$$

__Punto 3:__    
$m_3, \; T_3, \; P_3 \qquad \rightarrow \qquad h_3, \; s_3$

__Punto 4:__
$$h_4 = \frac{m_2 h_2 + m_3 h_3}{m_4}$$

$P_4, \; h_4 \qquad \rightarrow \qquad T_4, \; h'_4, \; h''_4, \; s'_4, \; s''_4$

$$x_4 = \frac{h_4 - h'_4}{h''_4 - h'_4}$$

$$s_4 = x_4 (s''_4 - s'_4) + s'_4$$

---
### b) Potencia Térmica Q

__Punto 5:__    
$P_5, \; t_5 \qquad \rightarrow \qquad h_5$
$$m_5 = m_4$$

Intercambiador de calor
$$Q_f = m_4 (h_5 - h_4)$$

---
### c) Potencia de la Turbina

__Punto 6:__    
$P_6, \; x_6 \qquad \rightarrow \qquad h'_6, \; h''_6, \; T_6, \; s'_6, \; s''_6$
$$h_6 = x_6(h''_6 - h'_6) + h'_6$$

$$s_6 = x_6 (s''_6 - s'_6) + s'_6$$

Potencia en la turbina
$$W = m_5 (h_5 - h_6)$$

---
### d) Variación de Entropía del Universo

$$\Delta S_{16} = m_1(s_6 - s_1)$$

$$\Delta S_{36} = m_3 (s_6 - s_3)$$

$$\Delta S_f = \frac{Q_f}{T_f}$$

Entonces
$$\Delta S_u = \Delta S_{16} + \Delta S_{36} + \Delta S_f$$

---
### e) Trabajo perdido
$$W_{perdido} = T_0 \Delta S_u$$

---
### Python

```python
# Punto 1
m_1 = 0.5 # [kg/s]
P_1 = 40e2 # [kPa]
x_1 = 0.7

hp_1 = 1087.40  ;  hpp_1 = 2800.34 # [kJ/kg]
sp_1 = 2.79652  ;  spp_1 = 6.06861 # [kJ/(kg·K)]

h_1 = x_1*(hpp_1 - hp_1) + hp_1
s_1 = x_1*(spp_1 - sp_1) + sp_1
```

```python
# Punto 2
P_2 = 15e2 # [kPa]
h_2 = h_1  # (válvula)
m_2 = m_1

T_2 = 198.29 + 273.15 # [K]
hp_2 = 844.66  ;  hpp_2 = 2789.89 # [kJ/kg]
sp_2 = 2.31447 ;  spp_2 = 6.44059 # [kJ/kg]

x_2 = (h_2 - hp_2)/(hpp_2 - hp_2)
s_2 = x_2*(spp_2 - sp_2) + sp_2
```

```python
# Punto 3
m_3 = 0.6 # [kg/s]
T_3 = 250 + 273.15 # [K]

# Cámara de mezcla
P_4 = P_3 = P_2
m_4 = m_2 + m_3

h_3 = 2923.52 # [kJ/kg]
s_3 = 6.70990 # [kJ/(kg·K)]
```

```python
# Punto 4
h_4 = (m_2*h_2 + m_3*h_3)/m_4

T_4 = 198.29 + 273.15 # [K]
hp_4 = 844.66  ;  hpp_4 = 2789.89 # [kJ/kg]
sp_4 = 2.31447 ;  spp_4 = 6.44059 # [kJ/kg]

x_4 = (h_4 - hp_4)/(hpp_4 - hp_4)

s_4 = x_4*(spp_4 - sp_4) + sp_4
```

```python
# Punto 5
P_5 = 14e2 # [kPa]
T_5 = 300 + 273.15 # [K]
m_5 = m_4
h_5 = 3041.62 # [kJ/kg]
s_5 = 6.95609 # [kJ/(kg·K)]

# Intercambiador de calor
Q = m_4*(h_5 - h_4)
```

```python
# Punto 6
P_6 = 0.5e2 # [kPa]
x_6 = 0.95

hp_6 = 340.56  ; hpp_6 = 2645.99 # [kJ/kg]
sp_6 = 1.09121 ; spp_6 = 7.59473 # [kJ/(kg·K)]
T_6 = 81.35 + 273.15 # [K]

h_6 = x_6*(hpp_6 - hp_6) + hp_6
s_6 = x_6*(spp_6 - sp_6) + sp_6

# Potencia en la turbina
W = m_5*(h_5 - h_6)
```

```python
# Variación de entropía
T_f = 800 # [K]
ΔS_16 = m_1*(s_6 - s_1)
ΔS_36 = m_3*(s_6 - s_3)
ΔS_f = Q/T_f

ΔS_u = ΔS_16 + ΔS_36 + (-ΔS_f)
```

```python
# Trabajo perdido
T_0 = 300 # [K]
W_perd = T_0*ΔS_u
```

```python
#@title Resultados
print("a) Estado de vapor en 2 y 4")
print("P_2 = %.0f [kPa]"%P_2)
print("T_2 = %.2f [K]"%T_2)
print("x_2 = %.3f"%x_2)
print("\nP_4 = %.0f [kPa]"%P_4)
print("T_4 = %.2f [K]"%T_4)
print("x_4 = %.3f"%x_4)
print("\nb) Potencia Térmica Q")
print("Q = %.3f [kW]"%Q)
print("\nc) Potencia de la turbina")
print("W = %.3f [kW]"%W)
print("\nd)Variación de entropía del universo")
print("ΔS_u = %.5f [kJ/(s·K)]"%ΔS_u)
print("\ne) Trabajo perdido")
print("W_perd = %.3f [kW]"%W_perd)
```

```python
import CoolProp.CoolProp as cp
```

```python
# Datos:
P_1 = 40e5 # [Pa]
x_1 = 0.7
m_1 = 0.5 # [kg/s]
P_2 = 15e5 # [Pa]
m_3 = 0.6 # [kg/s]
T_3 = 250 + 273.15 # [K]
P_5 = 15e5 # [Pa]
T_5 = 300 + 273.15 # [K]
P_6 = 0.5e5 # [Pa]
x_6 = 0.95
T_0 = 300 # [K]
P_0 = 1e5 # [Pa]
T_f = 800 # [K]
```

```python
# Punto 1
T_1 = cp.PropsSI('T','P',P_1,'Q',x_1,'Water')
h_1 = cp.PropsSI('H','P',P_1,'Q',x_1,'Water')
s_1 = cp.PropsSI('S','P',P_1,'Q',x_1,'Water')
```

```python
# Punto 2
m_2 = m_1
h_2 = h_1
x_2 = cp.PropsSI('Q','P',P_2,'H',h_2,'Water')
T_2 = cp.PropsSI('T','P',P_2,'H',h_2,'Water')
```

```python
# Cámara de mezcla
m_4 = m_2 + m_3
P_4 = P_3 = P_2
```

```python
# Punto 3
h_3 = cp.PropsSI('H','P',P_3,'T',T_3,'Water')
s_3 = cp.PropsSI('S','P',P_3,'T',T_3,'Water')
```

```python
cp.PhaseSI('P',P_3,'T',T_3,'Water')
```

```python
# Punto 4
h_4 = (m_2*h_2 + m_3*h_3)/m_4
```

```python
cp.PhaseSI('P',P_4,'H',h_4,'Water')
```

```python
T_4 = cp.PropsSI('T','P',P_4,'H',h_4,'Water')
x_4 = cp.PropsSI('Q','P',P_4,'H',h_4,'Water')
s_4 = cp.PropsSI('S','P',P_4,'H',h_4,'Water')
```

```python
# Punto 5
h_5 = cp.PropsSI('H','P',P_5,'T',T_5,'Water')
s_5 = cp.PropsSI('S','P',P_5,'T',T_5,'Water')
m_5 = m_4
# Intercambiador de calor
Q = m_4*(h_5 - h_4)
```

```python
# Potencia de la turbina
h_6 = cp.PropsSI('H','P',P_6,'Q',x_6,'Water')
s_6 = cp.PropsSI('S','P',P_6,'Q',x_6,'Water')

W = m_5*(h_5 - h_6)
```

```python
# Variación de entropía del universo
ΔS_16 = m_1*(s_6 - s_1)
ΔS_36 = m_3*(s_6 - s_3)
ΔS_f = Q/T_f

ΔS_u = ΔS_16 + ΔS_36 + (-ΔS_f)
```

```python
# Trabajo perdido
W_perd = T_0*ΔS_u
```

```python
#@title Resultados
print("a) Estado de vapor en 2 y 4")
print("P_2 = %.0f [kPa]"%(P_2/1000))
print("T_2 = %.2f [K]"%T_2)
print("x_2 = %.3f"%x_2)
print("\nP_4 = %.0f [kPa]"%(P_4/1000))
print("T_4 = %.2f [K]"%T_4)
print("x_4 = %.3f"%x_4)
print("\nb) Potencia Térmica Q")
print("Q = %.3f [kW]"%(Q/1000))
print("\nc) Potencia de la turbina")
print("W = %.3f [kW]"%(W/1000))
print("\nd)Variación de entropía del universo")
print("ΔS_u = %.5f [kJ/(s·K)]"%(ΔS_u/1000))
print("\ne) Trabajo perdido")
print("W_perd = %.3f [kW]"%(W_perd/1000))
```
