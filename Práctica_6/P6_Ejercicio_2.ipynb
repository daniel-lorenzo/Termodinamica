# Ejercicio 2
Proceso con vapor de agua. La variación de entropía entre la entrada y la salida de la turbina es de 3,2 kW/K. Determinar:
1. $\dot{W}$
2. $\eta_{iso}$

image.png

## Solución

Punto 1
  
$P_1 = 22 \; \mathrm{bar}$    
$T_1 = 300^\circ \mathrm{C}$
$$\begin{array}{l}
h_1 = 3019,29 \; \mathrm{kJ/kg} \\
s_1 = 6,71789 \; \mathrm{kJ/(kg \cdot K)}
\end{array}$$

Como $\Delta Su_{1-2} = 3,2 \; \mathrm{kJ/(kg \cdot K)}$

$$\Delta Su_{1-2} = \dot{m} (s_2 - s_1)$$

Entonces
$$s_2 = \frac{\Delta Su_{1-2}}{\dot{m}} + s_1$$

$$s_2 = \mathrm{\frac{3,2 \; kJ/(kg \cdot K)}{4 \; kg/s} + 6,71789 \; kJ/(kg \cdot K)}$$

$$s_2 = 7,518 \; \mathrm{kJ/(kg \cdot K)}$$
