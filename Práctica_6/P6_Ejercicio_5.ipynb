# Ejercicio 5
Determinar:

a. $\dot{W}$    
b. Estado del vapor en punto 4 (P, T, x)    
b. $\Delta S_u$

## Solución

### a) $W$

__Punto 1:__   
$P_1, \;T_1 \qquad \rightarrow \qquad h_1, \; s_1$

Para una turbina isoentrópica
$$s_{2rev} = s_1$$

__Punto 2:__    
Para una cámara de mezcla
$$P_2 = P_3 = P_4$$   
$P_2 \qquad \rightarrow \qquad h'_2, \; h''_2, \; s'_2, \; s''_2$

$$x_{2rev} = \frac{s_{2rev} - s'_2}{s''_2 - s'_2}$$

$$h_{2rev} = x_{2rev} (h''_2 - h'_2) + h'_2$$

Rendimiento isoentrópico (turbina)
$$\eta_{iso} = \frac{W}{W_{rev}} = \frac{h_1 - h_2}{h_1 - h_{2rev}}$$

Despejamos $h_2$
$$h_2 = \eta_{iso} (h_{2rev} - h_1) + h_1$$

Potencia real de la turbina
$$\dot{W} = \dot{m} (h_1 - h_2)$$

---
### b) Estado del vapor en el punto 4 $(P, \; T, \; x)$

Para una cámara de mezcla
$$m_4 = m_2 + m_3$$

$$ m_4 h_4 = m_2 h_2 + m_3 h_3$$

$P_3, \; T_3 \qquad \rightarrow \qquad h_3, \; s_3$

Despejamos $h_4$
$$h_4 = \frac{m_2 h_2 + m_3 h_3}{m_4}$$

$P_4, \; h_4 \qquad \rightarrow \qquad T_4, \; x_4 \; s_4$

---
### c) $\Delta S_u$

$$\Delta S_{14} = m_1 (s_4 - s_1)$$

$$\Delta S_{34} = m_3 (s_4 - s_3)$$

Variación de entropía del universo
$$\Delta S_u = \Delta S_{14} + \Delta S_{34}$$

```python
# Datos:
m_1 = 1.25 # [kg/s]
T_1 = 250 + 273.15 # [K]
P_1 = 10.5e5 # [Pa]
η_iso = 0.8
P_4 = 2e5 # [Pa]
m_3 = 3 # [kg/s]
T_3 = 240 + 273.15 # [K]
T_0 = 300 # [K]
P_0 = 1e5 # [Pa]
```

```python
# Punto 1
h_1 = 2941.12 # [kJ/kg]
s_1 = 6.90063 # [kJ/kg]

s_2rev = s_1
```

```python
# Punto 2: P_2=2 bar
hp_2 = 504.70    ;  hpp_2 = 2706.29  # [kJ/kg]
sp_2 = 1.53008   ;  spp_2 = 7.12683  # [kJ/kg]

x_2rev = (s_2rev - sp_2)/(spp_2 - sp_2)
h_2rev = x_2rev*(hpp_2 - hp_2) + hp_2
h_2 = η_iso*(h_2rev - h_1) + h_1
```

```python
W_rev = m_1*(h_1 - h_2rev)
W = η_iso*W_rev
```

```python
m_2 = m_1
m_4 = m_2 + m_3
```

```python
# Punto 3: P_3=2 bar; T_3=240°C
h_3 = 2951.10 # [kJ/kg]
s_3 = 7.67071 # [kJ/(kg·K)]
```

```python
h_4 = (m_2*h_2 + m_3*h_3)/m_4
```

```python
# P_4, h_4
t_4 = 200 # [°C]
s_4 = 7.50716 # [kJ/(kg·K)]
```

```python
ΔS_14 = m_1*(s_4 - s_1)
ΔS_34 = m_3*(s_4 - s_3)

ΔS_u = ΔS_14 + ΔS_34
```
