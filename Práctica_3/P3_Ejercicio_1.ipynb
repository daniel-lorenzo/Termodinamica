# Ejercicio 1
Un sistema cerrado se somete a un ciclo completo de dos procesos. En el primero se transfieren 60kJ de energía térmica al sistema, y éste entrega al medio un trabajo de 80kJ. En el segundo proceso, el medio exterior entrega al sistema un trabajo de 45kJ.

Se pide determinar:
1. El intercambio de calor durante el segundo proceso.
2. El trabajo y el calor intercambiado durante el ciclo.

image.png

Para determinar los puntos a) y b) deben plantear el primer principio para sistemas cerrados, hallar la variación de U, tener en cuenta que variación U12 = -variación U21, calcular el valor de Q21 y hacer la sumatoria de calores y trabajos.

## Solución

Para determinar los puntos a) y b) deben plantear el primer principio para sistemas cerrados, hallar la variación de U, tener en cuenta que $\Delta U_{12}$ = $-\Delta U_{21}$, calcular el valor de $Q_{21}$ y hacer la sumatoria de calores y trabajos.

image.png

### a)
$$Q_{12} - W_{12} = \Delta U_{12}$$

$$\mathrm{60 \; kJ - 80 \; kJ} = \Delta U_{12}$$

$$-20 \; \mathrm{kJ} = \Delta U_{12}$$

Entonces
$$\Delta U_{21} = - \Delta U_{12}$$

$$Q_{21} - \mathrm{(-45 \; kJ) = 20 \; kJ}$$

$$Q_{21} = -25 \; \mathrm{kJ}$$

### b)
$$Q = \mathrm{60 \; - 25 \; kJ = 35 \; kJ}$$

$$W = \mathrm{80 \; kJ - 45 \; kJ = 35 \; kJ}$$
