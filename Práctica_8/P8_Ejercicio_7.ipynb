## Solución

```python
# Datos:
P_0 = 1e5 # [Pa]
T_0 = 300 # [K]
P_1 = 20e5 # [Pa]
T_1 = 400 + 273 # [K]
m_2 = 3 # [kg/s]
x_2 = 1
P_3 = 1e5 # [Pa]
P_2 = P_4 = P_5 = P_6 = 10e5 # [Pa]
T_5 = 200 + 273.15 # [K]
x_6 = 0.8
P_7 = 1.5e5 # [Pa]
T_7 = 27 + 273 # [K]
T_8 = 87 + 273 # [K]
P_8 = 1.48e5 # [Pa]
T_9 = T_0
P_9 = P_0
P_10 = 3e5 # [Pa]
R_aire = 0.2870e3 # [J/(kg·K)]
cp_aire = 1.0047e3 # [J/(kg·K)]
η_isoT = 0.9
η_isoC = 0.8
W_T = 8000e3 # [W]
W_C = -W_T
```

```python
import CoolProp.CoolProp as cp
import numpy as np
```

```python
# Punto 1: P_1 = 20 bar ; T_1 = 400°C
h_1 = cp.PropsSI('H','P',P_1,'T',T_1,'Water')
s_1 = cp.PropsSI('S','P',P_1,'T',T_1,'Water')
# Punto 2: P_2 = 10 bar ; x_2 = 1
h_2 = cp.PropsSI('H','P',P_2,'Q',x_2,'Water')
s_2 = cp.PropsSI('S','P',P_2,'Q',x_2,'Water')
# Punto 3: P_3 = 1 bar => h', h'', s', s''
hp_3 = cp.PropsSI('H','P',P_3,'Q',0,'Water')
hpp_3 = cp.PropsSI('H','P',P_3,'Q',1,'Water')
sp_3 = cp.PropsSI('S','P',P_3,'Q',0,'Water')
spp_3 = cp.PropsSI('S','P',P_3,'Q',1,'Water')
# Punto 4: P_4 = 10 bar => h', h''
hp_4 = cp.PropsSI('H','P',P_4,'Q',0,'Water')
hpp_4 = cp.PropsSI('H','P',P_4,'Q',1,'Water')
# Punto 5: P_5 = 10 bar ; T_5 = 200°C
h_5 = cp.PropsSI('H','P',P_5,'T',T_5,'Water')
s_5 = cp.PropsSI('S','P',P_5,'T',T_5,'Water')
# Punto 6: P_6 = 10 bar ; x_6 = 0.8 => h', h'', s', s''
hp_6 = cp.PropsSI('H','P',P_6,'Q',0,'Water')  # h'
hpp_6 = cp.PropsSI('H','P',P_6,'Q',1,'Water') # h''
sp_6 = cp.PropsSI('S','P',P_6,'Q',0,'Water')  # s'
spp_6 = cp.PropsSI('S','P',P_6,'Q',1,'Water') # s''

h_6 = x_6*(hpp_6 - hp_6) + hp_6
s_6 = x_6*(spp_6 - sp_6) + sp_6
# Punto 7
h_7 = cp_aire*T_7
# Punto 8
h_8 = cp_aire*T_8
```

```python
#@title Resultados
print("Punto 1:")
print("h_1 = %.2f [kJ/kg]"%(h_1/1e3))
print("s_1 = %.5f [kJ/(kg·K)]"%(s_1/1e3))
print("Punto 2:")
print("h_2 = %.2f [kJ/kg]"%(h_2/1e3))
print("s_2 = %.5f [kJ/(kg·K)]"%(s_2/1e3))
print("Punto 3:")
print("hp_3 = %.2f [kJ/kg]"%(hp_3/1e3))
print("hpp_3 = %.2f [kJ/kg]"%(hpp_3/1e3))
print("Punto 4:")
print("hp_4 = %.2f [kJ/kg]"%(hp_4/1e3))
print("hpp_4 = %.2f [kJ/kg]"%(hpp_4/1e3))
print("Punto 5:")
print("h_5 = %.2f [kJ/kg]"%(h_5/1e3))
print("s_5 = %.5f [kJ/(kg·K)]"%(s_5/1e3))
print("Punto 6:")
print("h_6 = %.2f [kJ/kg]"%(h_6/1e3))
print("s_6 = %.5f [kJ/(kg·K)]"%(s_6/1e3))
print("Punto 7:")
print("h_7 = %.3f [kJ/kg]"%(h_7/1e3))
print("Punto 8:")
print("h_8 = %.3f [kJ/kg]"%(h_8/1e3))
```

### Turbina

$P_1, \; T_1 \qquad \rightarrow \qquad h_1, \; s_1$
$$s_{4rev} = s_1$$

Rendimiento isoentrópico
$$\eta_{isoT} = \frac{h_1 - h_4}{h_1 - h_{4rev}}$$

Despejo $h_4$
$$h_4 = h_1 - \eta_{iso} (h_1 - h_{4rev})$$

$P_4, \; h_4 \qquad \rightarrow \qquad s_4$

$P_3$    
$s_{3rev} = s_1$

Calidad de vapor ideal en 3
$$x_{3rev} = \frac{s_{3rev} - s'_3}{s''_3 - s'_3}$$

Entalpía reversible
$$h_{3rev} = x_{3rev} (h''_3 - h'_3) + h'_3$$

Rendimiento isoentrópico
$$\eta_{isoT} = \frac{h_1 - h_3}{h_1 - h_{3rev}}$$

Despejamos $h_3$
$$h_3 = h_1 - \eta_{iso} (h_1 - h_{3rev})$$

Calidad de vapor en 3
$$x_3 = \frac{h_3 - h'_3}{h''_3 - h'_3}$$

Entropía en 3
$$s_3 = x_3 (s''_3 - s'_3) + s'_3$$

```python
# Entropía reversible o ideal en 4
s_4rev = s_1
# Entalpía reversible o ideal en 4
h_4rev = cp.PropsSI('H','P',P_4,'S',s_4rev,'Water')
```

```python
# Entalpía en 4
h_4 = h_1 - η_isoT*(h_1 - h_4rev)
# Entropía en 4
s_4 = cp.PropsSI('S','P',P_4,'H',h_4,'Water')
# Entropía reversible o ideal en 3
s_3rev = s_1
# Calidad de vapor ideal en 3
x_3rev = (s_3rev - sp_3)/(spp_3 - sp_3)
# Entalpía ideal o reversible en 3
h_3rev = x_3rev*(hpp_3 - hp_3) + hp_3
# Entalpía en 3
h_3 = h_1 - η_isoT*(h_1 - h_3rev)
# Calidad de vapor en 3
x_3 = (h_3 - hp_3)/(hpp_3 - hp_3)
# Entropia en 3
s_3 = x_3*(spp_3 - sp_3) + sp_3
```

```python
#@title Resultados
print("s_4rev = %.5f [kJ/(kg·K)]"%(s_4rev/1e3))
print("h_4rev = %.2f [kJ/kg]"%(h_4rev/1e3))
print("h_4 = %.2f [kJ/kg]"%(h_4/1e3))
print("s_4 = %.5f [kJ/(kg·K)]"%(s_4/1e3))
print("s_3rev = %.5f [kJ/(kg·K)]"%(s_3rev/1e3))
print("x_3rev = %.2f"%x_3rev)
print("h_3rev = %.2f [kJ/kg]"%(h_3rev/1e3))
print("h_3 = %.2f [kJ/kg]"%(h_3/1e3))
print("x_3 = %.2f"%x_3)
print("s_3 = %.5f [kJ/(kg·K)]"%(s_3/1e3))
```

### Cámara de mezcla

Para la cámara de mezcla, tenemos:
$$m_2 h_2 + m_4 h_4 = m_5 h_5 \tag{1}$$
$$m_2 + m_4 = m_5 \tag{2}$$

Reemplazamos (2) en (1):
$$m_2 h_2 + m_4 h_4 = (m_2 + m_4) h_5$$

Despejamos $m_4$
$$m_4 = \frac{m_2 (h_2 - h_5)}{h_5 - h_4}$$

```python
# Flujo másico en 4
m_4 = m_2*(h_2 - h_5)/(h_5 - h_4)
```

```python
#@title Resultado
print("m_4 = %.3f [kg/s]"%m_4)
```

### Turbina

Primer principio termodinámica   
(Sistema abierto)

$$Q - W = \Delta H$$
$Q = 0 \;$, (se considera adiabático).

Desarrollamos
$$W = m_1 h_1 - (m_4 h_4 + m_3 h_3) \tag{3}$$

Tenemos que
$$m_1 = m_3 + m_4 \tag{4}$$

Reemplazamos (4) en (3)
$$W = (m_3 + m_4) h_1 - m_4 h_4 - m_3 h_3$$

Despejamos $m_3$
$$m_3 = \frac{W + m_4 (h_4 - h_1)}{h_1 - h_3}$$

```python
# Flujo másico en 3
m_3 = (W_T + m_4*(h_4 - h_1))/(h_1 - h_3)
# Flujo másico en 1
m_1 = m_3 + m_4
```

```python
#@title Resultados
print("m_3 = %.3f [kg/s]"%m_3)
print("m_1 = %.3f [kg/s]"%m_1)
```

### Intercambiador

Para el intercambiado de calor tenemos

$$m_5 h_5 + m_7 h_7 = m_6 h_6 + m_8 h_8$$
$$m_5 = m_6$$
$$m_7 = m_8$$

$$m_5 = m_4 + m_2$$

$$h_7 = c_{p,aire} T_7$$
$$h_8 = c_{p,aire} T_8$$

$$\Delta H_{6-5} + \Delta H_{8-7} = 0$$
$$m_5 (h_6 - h_5) + m_7 c_{p,aire}(T_8 - T_7) = 0$$

$$m_5 h_5 + m_7 h_7 = m_5 h_6 + m_7 h_8$$

Flujo másico (aire) en 7 - 8

$$m_7 = \frac{m_5 (h_5 - h_6)}{c_{p,aire}(T_8 - T_7)} = \frac{m_5 (h_5 - h_6)}{h_8 - h_7}$$

```python
m_5 = m_6 = m_2 + m_4
h_7 = cp_aire*T_7
h_8 = cp_aire*T_8

# Flujo másico en 7-8
m_7 = (m_5*(h_5 - h_6))/(h_8 - h_7)
```

```python
#@title Resultado
print("m_7 = %.3f [kg/s]"%m_7)
```

### Compresor

Como consideramos el caso isoentrópico

$$\Delta S_{c,rev} = m \left( c_{p,aire} \ln \frac{T_{10rev}}{T_9} - R \ln \frac{P_{10}}{P_9} \right) = 0$$

Despejamos $T_{10rev}$

$$\ln \frac{T_{10rev}}{T_9} = \frac{R}{c_p} \ln \frac{P_{10}}{P_9}$$

$$T_{10rev} = T_9 e^{R/c_p \ln \frac{P_{10}}{P_9}}$$

Rendimiento isoentrópico
$$\eta_{iso} = \frac{W_{rev}}{W} = \frac{m c_p (T_9 - T_{10rev})}{m c_p (T_9 - T_{10})}$$

Despejamos $T_{10}$

$$T_{10} = T_9 - \frac{T_9 - T_{10rev}}{\eta_{isoC}}$$

Primer principio termodinámica
$$Q - W = \Delta H$$

$Q=0$
$$-W_C = m_9 c_p (T_{10} - T_9)$$

Despejamos $m_9$
$$m_9 = \frac{-W_C}{c_p (T_{10} - T_9)}$$

```python
T_10rev = T_9*np.exp(R_aire/cp_aire*np.log(P_10/P_9) )
T_10 = T_9 - (T_9 - T_10rev)/(η_isoC)
m_9 = -W_C/(cp_aire*(T_10 - T_9))
```

```python
#@title Resultados
print("T_10rev = %.2f [K]"%T_10rev)
print("T_10 = %.2f [K]"%T_10)
print("m_9 = %.3f [kg/s]"%m_9)
```

### A) Cálculo variación de entropías

$$\Delta S_{1-3} = m_3 (s_3 - s_1)$$
$$\Delta S_{1-6} = m_4 (s_6 - s_1)$$
$$\Delta S_{2-6} = m_2 (s_6 - s_2)$$

$$\Delta S_{7-8} = m_7 \left[ c_p \ln \frac{T_8}{T_7} - R \ln \frac{P_8}{P_7} \right]$$

$$\Delta S_{9-10} = m_9 \left[ c_p \ln \frac{T_{10}}{T_9} - R \ln \frac{P_{10}}{P_9} \right]$$

$$\Delta S_u = \Delta S_{1-3} + \Delta S_{1-6} + \Delta S_{2-6} + \Delta S_{7-8} + \Delta S_{9-10}$$

```python
ΔS_13 = m_3*(s_3 - s_1)
ΔS_16 = m_4*(s_6 - s_1)
ΔS_26 = m_2*(s_6 - s_2)
ΔS_78 = m_7*(cp_aire*np.log(T_8/T_7) - R_aire*np.log(P_8/P_7))
ΔS_910 = m_9*(cp_aire*np.log(T_10/T_9) - R_aire*np.log(P_10/P_9))
# Variación de entropía del universo
ΔS_u = ΔS_13 + ΔS_16 + ΔS_26 + ΔS_78 + ΔS_910
```

```python
#@title Resultados
print("ΔS_13 = %.3f [kJ/K]"%(ΔS_13/1000))
print("ΔS_16 = %.3f [kJ/K]"%(ΔS_16/1000))
print("ΔS_26 = %.3f [kJ/K]"%(ΔS_26/1000))
print("ΔS_78 = %.3f [kJ/K]"%(ΔS_78/1000))
print("ΔS_910 = %.3f [kJ/K]"%(ΔS_910/1000))
print("\nΔS_u = %.3f [kJ/K]"%(ΔS_u/1000))
```

### B) Cálculo rendimiento exergético

$$\Delta Ex_{1-3} = m_3 (h_3 - h_1) - T_0 \Delta S_{1-3}$$
$$\Delta Ex_{1-6} = m_4 (h_6 - h_1) - T_0 \Delta S_{1-6}$$
$$\Delta Ex_{2-6} = m_2 (h_6 - h_2) - T_0 \Delta S_{2-6}$$
$$\Delta Ex_{7-8} = m_7 c_p(T_8 - T_7) - T_0 \Delta S_{7-8}$$
$$\Delta Ex_{9-10} = m_9 c_p (T_{10} - T_9) - T_0 \Delta S_{9-10}$$

Rendimiento exergético
$$\eta_{ex} = \frac{|\Delta Ex_{9-10}| + |\Delta Ex_{7-8}|}{|\Delta Ex_{1-3}| + |\Delta Ex_{1-6}| + |\Delta Ex_{2-6}|}$$

```python
ΔEx_13 = m_3*(h_3 - h_1) - T_0*ΔS_13
ΔEx_16 = m_4*(h_6 - h_1) - T_0*ΔS_16
ΔEx_26 = m_2*(h_6 - h_2) - T_0*ΔS_26
ΔEx_78 = m_7*cp_aire*(T_8 - T_7) - T_0*ΔS_78
ΔEx_910 = m_9*cp_aire*(T_10 - T_9) - T_0*ΔS_910

η_ex = (ΔEx_910 + ΔEx_78)/abs(ΔEx_13 + ΔEx_16 + ΔEx_26)
```

```python
#@title Resultados
print("ΔEx_13 = %.3f [kJ]"%(ΔEx_13/1000))
print("ΔEx_16 = %.3f [kJ]"%(ΔEx_16/1000))
print("ΔEx_26 = %.3f [kJ]"%(ΔEx_26/1000))
print("ΔEx_78 = %.3f [kJ]"%(ΔEx_78/1000))
print("ΔEx_910 = %.3f [kJ]"%(ΔEx_910/1000))
print("\nη_ex = %.3f"%η_ex)
```

---
### C) Cálculo rendimiento exergético de la turbina

$$\Delta S_{1-4} = m_4 (s_4 - s_1)$$

$$\Delta Ex_{1-4} = m_4 (h_4 - h_1) - T_0 \Delta S_{1-4}$$

$$\eta_{ex,Turb} = \frac{|W_T|}{|\Delta Ex_{1-3}| + |\Delta Ex_{1-4}|}$$

```python
ΔS_14 = m_4*(s_4 - s_1)
ΔEx_14 = m_4*(h_4 - h_1) - T_0*ΔS_14
# Rendimiento exergético de la turbina
η_exTurb = W_T/abs(ΔEx_13 + ΔEx_14)
```

```python
#@title Resultados
print("ΔS_14 = %.4f [kJ/K]"%(ΔS_14/1000))
print("ΔEx_14 = %.3f [kJ]"%(ΔEx_14/1000))
print("\nη_exTurb = %.3f"%η_exTurb)
```

---
### D) Cálculo rendimiento exergético del compresor

$$\eta_{ex,Comp} = \frac{|\Delta Ex_{9-10}|}{|W_C|}$$

```python
# Rendimiento exergético del compresor
η_exComp = ΔEx_910/abs(W_C)
```

```python
#@title Resultado
print("\nη_exComp = %.3f"%η_exComp)
```
