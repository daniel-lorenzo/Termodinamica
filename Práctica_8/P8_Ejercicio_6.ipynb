## Soluci√≥n

```python
# Datos:
m_1 = 13 # [kg/s]
P_1 = 6000 # [kPa]
T_1 = 400 + 273 # [K]
P_2 = 700 # [kPa]
x_2 = 1
T_F = 800 # [K]
T_3 = 400 + 273 # [K]
P_3 = P_2
P_4 = 5 # [kPa]
x_4 = 0.9
P_0 = 100 # [kPa]
T_0 = 290 # [K]
```

```python
import CoolProp.CoolProp as cp
```

```python
# Punto 1
h_1 = cp.PropsSI('H','P',P_1*1e3,'T',T_1,'Water')/1e3 # [kJ/kg]
s_1 = cp.PropsSI('S','P',P_1*1e3,'T',T_1,'Water')/1e3 # [kJ/(kg¬∑K)]
# Punto 2
h_2 = cp.PropsSI('H','P',P_2*1e3,'Q',x_2,'Water')/1e3
s_2 = cp.PropsSI('S','P',P_2*1e3,'Q',x_2,'Water')/1e3
# Punto 3
h_3 = cp.PropsSI('H','P',P_3*1e3,'T',T_3,'Water')/1e3
s_3 = cp.PropsSI('S','P',P_3*1e3,'T',T_3,'Water')/1e3
# Punto 4
h_4 = cp.PropsSI('H','P',P_4*1e3,'Q',x_4,'Water')/1e3
s_4 = cp.PropsSI('S','P',P_4*1e3,'Q',x_4,'Water')/1e3
```

```python
#@title Resultados:
print("h_1 = %.2f [kJ/kg]\ts_1 = %.4f [kJ/(kg¬∑K)]"%(h_1,s_1))
print("h_2 = %.2f [kJ/kg]\ts_2 = %.4f [kJ/(kg¬∑K)]"%(h_2,s_2))
print("h_3 = %.2f [kJ/kg]\ts_3 = %.4f [kJ/(kg¬∑K)]"%(h_3,s_3))
print("h_4 = %.2f [kJ/kg]\ts_4 = %.4f [kJ/(kg¬∑K)]"%(h_4,s_4))
```

Potencia en A
$$W_a = m (h_1 - h_2)$$

Potencia en B
$$W_b = m (h_3 - h_4)$$

Calor fuente
$$Q_{Fuente} = m (h_3 - h_2)$$

Potencia
$$W = W_a + W_b$$

Variaci√≥n de exerg√≠a 1-4
$$\Delta Ex_{14} = m[(h_4 - h_1) - T_0 (s_4 - s_1)]$$

Calor √∫til 2-3
$$Q_{u23} = Q_{Fuente} - T_0 \frac{Q_{Fuente}}{T_{fuente}}$$

Potencia consumida
$$W_{Consumido} = \Delta Ex_{14} + Q_{u23}$$

Rendimiento exerg√©tico
$$\eta_{ex} = \frac{W}{W_{Consumido}}$$

```python
m = m_1
# Potencia en A
W_a = m*(h_1 - h_2)
# Potencia en B
W_b = m*(h_3 - h_4)
# Calor fuente
Q_F = m*(h_3 - h_2)
# Potencia
W = W_a + W_b
# Variaci√≥n de exerg√≠a 1-4
ŒîEx_14 = m*((h_4 - h_1) - T_0*(s_4 - s_1))
# Calor √∫til 2-3
Q_u23 = Q_F - T_0*Q_F/T_F
# Trabajo consumido
W_cons = abs(ŒîEx_14) + abs(Q_u23)
# Rendimiento exerg√©tico
Œ∑_ex = W/W_cons
```

```python
#@title Resultados
print("W_a = %.2f [kW]"%W_a)
print("W_b = %.2f [kW]"%W_b)
print("Q_F = %.2f [kW]"%Q_F)
print("W = %.3f [kW]"%W)
print("ŒîEx_14 = %.3f [kW]"%ŒîEx_14)
print("Q_u23 = %.2f [kW]"%Q_u23)
print("W_cons = %.3f[kW]"%W_cons)
print("\nŒ∑_ex = %.3f"%Œ∑_ex)
```
