$P_0 = 1 \; \mathrm{bar}$

## Solucin

```python
#@title Datos:
P_0 = 1e5           # [Pa]
P_1 = 10e5          # [Pa]
P_2 = P_1
P_3 = 1e5           # [Pa]
P_4 = 1e5           # [Pa]
x_4 = 0
P_5 = 5e5           # [Pa]
P_p = 3e5           # [Pa]
T_0 =  27 + 273.15  #  [K]
T_1 = 180 + 273.15  # [K]
T_2 = 510 + 273.15  # [K]
T_i = 40 + 273.15   # [K]
c   = 83.6e3        # [J/K]
T_F1 = 700          # [K]
T_F2 = 10 + 273.15  # [K]
Q_2 = -4180e3       # [J]
W_perd = 209e3      # [J]
_iso = 0.7
m_v = 100           # [kg]
x_c = 0.9
```

### Mquina frigorfica

Trabajo perdido
$$W_{perd} = T_0 \Delta S_{MF}$$

Variacin de entropa mquina frigorfica
$$\Delta S_{MF} = \frac{W_{perd}}{T_0}$$

Variacin de entropa fuente 2:
$$\Delta S_{F2} = \frac{Q_2}{T_2}$$

Luego
$$W_{perd} = T_0 \left[ \frac{Q_2}{T_2} + c \ln \left( \frac{T_f}{T_i} \right) \right]$$

Despejamos $T_f$
$$\frac{W_{perd}}{c T_0} - \frac{Q_2}{c T_2} = \ln \frac{T_f}{T_i}$$

$$e^{W_{perd}/(cT_0) - Q_2/(cT_{F2})} = \frac{T_f}{T_i}$$

Temperatura final del cuerpo
$$T_f = T_i e^{W_{perd}/(cT_0) - Q_2/(cT_{F2})}$$

Variacin de entropa del cuerpo
$$\Delta S_{C1} = c  \ln \left( \frac{T_f}{T_i} \right)$$

Calor 1
$$Q_1 = c (T_f - T_i)$$

Clculo trabajo turbina
$$Q_1 = 0,6 W_t + Q_2$$

Trabajo turbina
$$W_t = \frac{Q_1 - Q_2}{0,6}$$

```python
import numpy as np
```

```python
#@title Clculo mquina frigorfica
# Temperatura final
T_f = T_i*np.exp( W_perd/(c*T_0) - Q_2/(c*T_F2) )
# Calor 1
Q_1 = c*(T_f - T_i)
# Trabajo turbina
W_t = (abs(Q_1) - abs(Q_2))/0.6
# Variacin de entropa fuente 2
S_F2 = Q_2/T_F2
# Variacin de entropa cuerpo
S_C1 = c*np.log(T_f/T_i)
# Variacin de entropa
S_MF = W_perd/T_0
# Calor til 1
Q_u1 = Q_1 - T_0*S_C1
# Calor til 2
Q_u2 = Q_2 - T_0*S_F2
```

```python
#@title Resultados mquina frigorfica
print("Temperatura final")
print("T_f = %.2f [K]\n"%T_f)
print("Variacin de entropa")
print("S_MF = %.4f [kJ/K]\n"%(S_MF/1000))
print("Calor 1")
print("Q_1 = %.3f [kJ]\n"%(Q_1/1000))
print("Trabajo turbina")
print("W_t = %.3f [kJ/s]\n"%(W_t/1000))
print("Variacin de entropa fuente 2")
print("S_F2 = %.4f [kJ/K]\n"%(S_F2/1000))
print("Variacin de entropa cuerpo")
print("S_C1 = %.3f [kJ/K]\n"%(S_C1/1000))
print("Calor til 1")
print("Q_u1 = %.2f [kJ]\n"%(Q_u1/1000))
print("Calor til 2")
print("Q_u2 = %.2f [kJ]\n"%(Q_u2/1000))
```

### Bomba

__Punto 4:__    
$$\left\{
\begin{array}{l}
P_4 = 1  \; \mathrm{bar} \\
x_4 = 0
\end{array}
\right. \qquad \rightarrow \qquad h_4,  \; s_4$$

__Punto 5:__
$$s_{5rev} = s_4$$

En tabla con $$P_5, \; s_{5rev} \qquad \rightarrow \qquad h_{5rev}$$

Rendimiento isoentrpico
$$\eta_{iso} = \frac{h_{5rev} - h_4}{h_5 - h_4}$$

Despejamos $h_5$
$$h_5 = h_4 + \frac{h_{5rev} - h_4}{\eta_{iso}}$$

En tabla con
$$P_5, \; h_5 \qquad \rightarrow \qquad s_5$$

Trabajo en la bomba
$$W_b = 0,15 W_t$$
$$-W_b = m_4 (h_5 - h_4)$$

Despejamos $m_4$
$$m_4 = \frac{-W_b}{h_5 - h_4}$$

Variacin de entropa
$$\Delta S_{45} = m_4 (s_5 - s_4)$$

Variacin de exerga
$$\Delta Ex_{45} = (h_5 - h_4) - T_0 \Delta S_{45}$$

```python
!pip install CoolProp
```

```python
import CoolProp.CoolProp as cp
```

```python
#@title Clculo bomba
# Punto 4
h_4 = cp.PropsSI('H','P',P_4,'Q',x_4,'Water')
s_4 = cp.PropsSI('S','P',P_4,'Q',x_4,'Water')
# Punto 5
s_5rev = s_4
h_5rev = cp.PropsSI('H','P',P_5,'S',s_5rev,'Water')
h_5 = h_4 + (h_5rev - h_4)/_iso
s_5 = cp.PropsSI('S','P',P_5,'H',h_5,'Water')
# Trabajo en la bomba
W_b = -0.15*W_t
# Flujo msico en 4
m_4 = -W_b/(h_5 - h_4)
# Variacin de entropa 4-5
S_45 = m_4*(s_5 - s_4)
# Variacin de exerga 4-5
Ex_45 = m_4*(h_5 - h_4) - T_0*S_45
```

```python
#@title Resultados
print("h_4 = %.2f [kJ/K]"%(h_4/1000))
print("s_4 = %.5f [kJ/K]"%(s_4/1000))
print("s_5rev = %.5f [kJ/kg]"%(s_5rev/1000))
print("h_5rev = %.2f [kJ/kg]"%(h_5rev/1000))
print("h_5 = %.2f [kJ/K]"%(h_5/1000))
print("s_5 = %.5f [kJ/K]"%(s_5/1000))
print("W_b = %.3f [kJ/s]"%(W_b/1000))
print("m_4 = %.2f [kg/s]"%m_4)
print("S_45 = %.3f [kJ/K]"%(S_45/1000))
print("Ex_45 = %.3f [kJ/K]"%(Ex_45/1000))
```

### Cilindro-Pistn

$$Q - W =  \Delta U$$

$Q=0$
$$-W_h - W_{\Delta V} = \Delta U$$

$$-W_h = W_{\Delta V} + \Delta U = \Delta H $$

$$-W_h = P m_v (v_f - v_i) + m_v (u_f - u_i)$$

$$-W_h = m_v (h_f - h_i)$$

$P_p, \; x_c \qquad \rightarrow$

$$h', h'' \rightarrow h_i$$
$$s', s'' \rightarrow s_i$$
$$v', v'' \rightarrow v_i$$
$$u', u'' \rightarrow u_i$$

$$-(0,2 W_t) = m_v (h_f - h_i)$$

$$h_f = \frac{-(0,2 W_t)}{m_v} + h_i$$

$$x = \frac{h_f - h'}{h'' - h'}$$

$P, \; x_f \qquad \rightarrow \qquad s_f, \; v_f, \; u_f$

$$\Delta S_{C-P} = m_v (s_f - s_i)$$

$$\Delta Ex_V = \Delta U - T_0 \Delta S_{C-P} + P_0 \Delta V$$

$$\Delta Ex_V = m_v (u_f - u_i) - T_0 \Delta S_{C-P} + P_0 [m_v (v_f - v_i)]$$

```python
#@title Cilindro-Pistn
h_i = cp.PropsSI('H','P',P_p,'Q',x_c,'Water')
s_i = cp.PropsSI('S','P',P_p,'Q',x_c,'Water')
v_i = 1/cp.PropsSI('D','P',P_p,'Q',x_c,'Water')
u_i = cp.PropsSI('U','P',P_p,'Q',x_c,'Water')

hp = cp.PropsSI('H','P',P_p,'Q',0,'Water')
hpp = cp.PropsSI('H','P',P_p,'Q',1,'Water')
```

```python
# Entalpa final del vapor
h_f = -(-0.2*W_t)/m_v + h_i
# Calidad final del vapor
x_cf = (h_f - hp)/(hpp - hp)
```

```python
# Entropia especfica final
s_f = cp.PropsSI('S','P',P_p,'Q',x_cf,'Water')
# Volumen especfico final
v_f = 1/cp.PropsSI('D','P',P_p,'Q',x_cf,'Water')
# Energan interna especfica final
u_f = cp.PropsSI('U','P',P_p,'Q',x_cf,'Water')
```

```python
# Variacin de entropa cilindro-pistn
S_cp = m_v*(s_f - s_i)
```

```python
# Variacin de exerga del vapor
Ex_v = m_v*(u_f - u_i) - T_0*S_cp + P_0*(v_f - v_i)
```

```python
# Trabajo por variacin de volumen
W_V = P_p*m_v*(v_f - v_i)
# Trabajo til
W_u = W_V - P_0*m_v*(v_f - v_i)
```

```python
#@title Resultados:
print("h_f = %.2f [kJ/K]"%(h_f/1000))
print("x_cf = %.3f"%x_cf)
print("s_f = %.2f [kJ/(kgK)]"%(s_f/1000))
print("v_f = %.3f [m^3/kg]"%v_f)
print("u_f = %.3f [kJ/kg]"%(u_f/1000))
print("S_cp = %.1f [kJ/kg]"%(S_cp/1000))
print("Ex_v = %.1f [kJ]"%(Ex_v/1000))
print("W_u = %.1f [kJ]"%(W_u/1000))
```

### Turbina

**Punto 1:**

$P_1, \; T_1 \qquad \rightarrow \qquad h_1, \; s_1$

**Punto 2:**

$P_2, \; T_2 \qquad \rightarrow \qquad h_2, \; s_2$

**Punto 3:**

$P_3, \; s_{3rev} = s_2 \qquad \rightarrow \qquad h_{3rev}$

Rendimiento isoentrpico
$$\eta_{iso} = \frac{h_2 - h_3}{h_2 - h_{3rev}}$$

Despejamos $h_3$
$$h_3 = \eta_{iso} (h_{3rev} - h_2) + h_2$$

Luego

$P_3, \; h_3 \qquad \rightarrow \qquad s_3$

$$-W_t = m_1 (h_3 - h_2)$$

Despejamos $m_1$
$$m_1 = \frac{W_t}{h_2 - h_3}$$

$$\Delta S_{13} = m_1 (s_3 - s_1)$$

$$\Delta Ex_{13} = m_1 (h_3 - h_1) - T_0 \Delta S_{13}$$

```python
# Punto 1: P_1, T_1
h_1 = cp.PropsSI('H','P',P_1,'T',T_1,'Water')
s_1 = cp.PropsSI('S','P',P_1,'T',T_1,'Water')
# Punto 2: P_2, T_2
h_2 = cp.PropsSI('H','P',P_2,'T',T_2,'Water')
s_2 = cp.PropsSI('S','P',P_2,'T',T_2,'Water')
# Punto 3: P_3, s_3rev = s_2
s_3rev = s_2
h_3rev = cp.PropsSI('H','P',P_3,'S',s_3rev,'Water')
h_3 = _iso*(h_3rev - h_2) + h_2
s_3 = cp.PropsSI('S','P',P_3,'H',h_3,'Water')
m_1 = W_t/(h_2 - h_3)
S_13 = m_1*(s_3 - s_1)
Ex_13 = m_1*(h_3 - h_1) - T_0*S_13
```

```python
#@title Resultados:
print("h_1 = %.2f [kJ/kg]"%(h_1/1000))
print("s_1 = %.5f [kJ/kg]"%(s_1/1000))
print("h_2 = %.2f [kJ/kg]"%(h_2/1000))
print("s_2 = %.5f [kJ/kg]"%(s_2/1000))
print("s_3rev = %.5f [kJ/kg]"%(s_3rev/1000))
print("h_3rev = %.2f [kJ/kg]"%(h_3rev/1000))
print("h_3 = %.2f [kJ/kg]"%(h_3/1000))
print("s_3 = %.5f [kJ/kg]"%(s_3/1000))
print("m_1 = %.2f [kg]"%m_1)
print("S_13 = %.3f [kJ/K]"%(S_13/1000))
print("Ex_13 = %.2f [kJ]"%(Ex_13/1000))
```

### Fuente

$$Q_c = m (h_2 - h_1)$$

$$\Delta S_{F1} = \frac{Q_c}{T_{F1}}$$

$$\Delta S_U = \Delta S_{C1} + \Delta S_{F2} + \Delta S_{45} + \Delta S_{C-P} + \Delta S_{13} + \Delta S_{F1}$$

```python
Q_c = m_1*(h_2 - h_1)
S_F1 = -Q_c/T_F1
S_u = S_C1 + S_F2 + S_45 + S_cp + S_13 + S_F1
```

```python
#@title Resultados:
print("Q_c = %.3f [kJ]"%(Q_c/1000))
print("S_F1 = %.3f [kJ/K]"%(S_F1/1000))
print("S_u = %.2f [kJ/K]"%(S_u/1000))
```
