# Ejemplo 1
Un dispositivo de cilindro-embolo adiab치tico, contiene una masa de 2 kg de vapor de agua, inicialmente saturado a una presi칩n de 2 atm. El cilindro est치 provisto con topes de manera tal que el volumen m치ximo que puede desarrollar es de 2 veces el volumen inicial ocupado por el vapor de agua. Si mediante un eje que acciona una rueda de paletas, se entrega un trabajo de 500 kJ al vapor contenido.     
__Plantear:__ el sistema de ecuaciones necesario para determinar el estado final del vapor.

## Soluci칩n
__Datos:__ 
$\begin{array}{|l|l|l|l|l|}
\hline
m & x_i & P_i & P_i=P & V_{max}=2V_i & W_{pal} \\
\hline
\end{array}$

__Incognitas:__
$\begin{array}{|l|l|}
\hline
P_f & h_f \\
\hline
\end{array}$

__Definici칩n:__
$$ \Delta H = \Delta U + W_{\Delta V} $$
__Primer principio__
$$ Q = \underbrace{ \Delta U + W_{\Delta V} }_{\Delta H} + W_{pal} $$
__Entonces__
$$ Q = \Delta H + W_{pal} $$

### Primer hip칩tesis
El 칠mbolo no llega al tope.

Condici칩n a cumplir: $V_f < 2 V_i$

__Determinar:__ $h_f$; $v_i$; $v_f$

$$\begin{array}{l|l|l|l}
 & \mathrm{Ecuaciones} & & \mathrm{Incognitas} \\
 \hline
1) & Q = \dot{m} (h_f - h_i) + W_{pal} &  & Q, \, h_f, \, h_i \\
2) & h_i = h''_i & f(P) \mbox{ vapor saturado } x_i = 1 \\
3) & Q = 0 & \mbox{porque es adiab치tico} \\
4) & V_i = \dot{m} v_i & & v_i, \, Vi \\
5) & v_i = v''_i & f(P) \, \mbox{vapor saturado } x_i = 1 \\
6) & V_f = \dot{m} v_f & & V_f, \, v_f \\
7) & v_f & f(h_f;P) \\
\hline
\end{array}$$

__7 ecuaciones - 7 incognitas__

Verif. condic. hip. $V_f < 2 V_i$

image.png

Como el $V_f$ aumenta a $P$ constante en el estado final el vapor es sobrecalentado

### Segunda hip칩tesis
$$ V_f = 2 V_i = V_{max} $$

debe cumplir: $P_f > P_i$

__Determinar:__ $V_f$, $v_f$, $u_f$

$$\begin{array}{l|l|l|l}
 & \mathrm{Ecuaciones} & & \mathrm{Incognitas} \\
\hline
1) & V_f = 2 V_i & & V_f \\
2) & V_f = m v_f & & v_f \\
3) & \Delta U = m (u_f - u_i) & & \Delta U, \, u_f, \, u_i \\
4) & u_i = u''_i & f(P) \; \mbox{vapor saturado} \\
5) & Q = \Delta U + W_{\Delta V} + W_{pal} & & Q, \, W_{\Delta V} \\
6) & Q = 0 & \mbox{porque es adiab치tico} \\
7) & W_{\Delta V} = P (V_f - V_i) \\
\hline
\end{array}$$

__7 ecuaciones - 7 incognitas__

image.png

Dado que:  
$\left.
\begin{array}{l}
P_f > P_i \\
V_f > V_i 
\end{array}
\right\}
\; \therefore \;$ es vapor sobrecalentado.

```python
# importa bilioteca CoolProp (tabla de propiedades termodin치micas)
import CoolProp.CoolProp as cp
```

```python
# Datos
m = 2 # kg
xi = 1
P = 2*101325 # Pa
Pi = P
Wpal = -500e3 # J

# Consultamos la entalp칤a inicial con [Pi, xi]
hi = cp.PropsSI('H','P',Pi,'Q',xi,'Water')

# Consultamos el volumen espec칤fico inicial con [Pi, xi]
vi = 1/cp.PropsSI('D','P',Pi,'Q',xi,'Water')

Q = 0 # porque es adiab치tico

# De la ecuaci칩n 1 despejamos hf
hf = hi - Wpal/m

# Consultamos en que fase se encuenta al final
fasef = cp.PhaseSI('P',P,'H',hf,'Water')

Vi = m*vi

# Consultamos el volumen espec칤fico final con [P, hf]
vf = 1/cp.PropsSI('D','P',P,'H',hf,'Water')

Vf = m*vf

# Imprime resultados
print('hi = %s kJ/kg'%round(hi/1000,2))
print('hf = %s kJ/kg'%round(hf/1000,2))
print('fase final = %s'%fasef)
print('Vi = %s m^3'%round(Vi,2))
print('Vf = %s m^3'%round(Vf,2))
print('Pf = %s kPa'%round(P/1000,3))
```
