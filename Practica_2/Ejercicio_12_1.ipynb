# Ejercicio 12.1
Un cilindro-pistón diatérmico contiene $CO_2$. Inicialmente el pistón que genera una presión de 10 bares. Se encuentra trabado. Determinar el intercambio de energía en el medio exterior cuando se quitan las trabas.

image.png

$\begin{array}{l}
\mathrm{Temp. \, ambiente = 20^\circ \; C}\\
P_0 = 1 \, \mathrm{bar}
\end{array}$

## Solución

```python
# Datos:
P_2 = 10e5 # Pa
P_1 = 1e5  # Pa
V_1 = 10 # m^3
T_amb = 20 + 273.15 # K
```

Como es diatérmico $T_1 = T_2 = T_{amb}$

$$ \frac{P_1 V_1}{T_1} = \frac{P_2 V_2}{T_2} $$

Entonces:
$$ V_2 = \frac{P_1 V_1}{P_2} $$

```python
V_2 = (P_1*V_1)/P_2
```

```python
#@title Resultado:
print('V_2 = %.2f m^3'%V_2)
```

Primer principio
$$ Q = \Delta U + W_{\Delta V} $$

$$ Q = m \, c_v \, \underbrace{(T_2 - T_1)}_{= \, 0} + P_p (V_2 - V_1) $$

Entonces:
$$ Q = W_{\Delta V} = P_p (V_2 - V_1) $$

El trabajo por variación de volumen $W_{\Delta V}$ se transforma en calor

```python
Q = P_2*(V_2 - V_1)
```

```python
#@title Resultado:
print('Q = %.1f kJ'%(Q/1000))
```
