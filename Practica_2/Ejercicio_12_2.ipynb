# Ejercicio 12.2
Un cilindro pistón adiabático contiene $CO_2$. Inicialmente, el pistón que genera una presión de 10 bares absoluta, se encuentra trabado. Determinar el intercambio de energía con el medio exterior cuando se quitan las trabas.

image.png

$\begin{array}{l}
\mathrm{Temp. \, ambiente = 20^\circ \; C}\\
P_0 = 1 \, \mathrm{bar}
\end{array}$

## Solución

```python
# Datos:
P_1 = 1e5 # Pa
P_2 = 10e5 # Pa
V_1 = 10 # m^3
T_1 = 20 + 273.15 # K
R = 0.1889e3 # J/(kg·K)
cv = 0.6552e3 # J/(kg·K)
```

Ecuación de estado
$$ PV = mRT $$

Despejamos $m$
$$ m = \frac{P_1 V_1}{R T_1} $$

```python
# Cálculo de la masa de CO_2
m = (P_1*V_1)/(R*T_1)
```

```python
#@title Resultado:
print('m = %.4f kg'%m)
```

Primer principio
$$ Q = \Delta U + W_{\Delta V} $$

Tenemos que:

$\left\{
\begin{array}{l}
Q = 0 \\
\Delta U = m \, c_v (T_2 - T_1) \\
W_{\Delta V} = P_2 (V_2 - V_1) \\
PV = m RT
\end{array}
\right.$

Reemplazamos:
$$ 0 = \Delta U + W_{\Delta V} $$

$$ 0 = m \, c_v (T_2 - T_1) + P_2 (V_2 - V_1) $$

$$ 0 = m \, c_v (T_2 - T_1) + m R T_2 - P_2 V_1 $$

Despejo $T_2$
$$ T_2 = \frac{T_1 c_v m + P_2 V_1}{m \, (c_v + R)} $$

```python
T_2 = (T_1*cv*m + P_2*V_1)/(m*(cv + R))
```

```python
#@title Resultado:
print('T_2 = %.2f K'%T_2)
```

```python
# Variación de energía interna
DU = m*cv*(T_2 - T_1)
# Trabajo por variación de volumen
W_DV = -DU
```

```python
#@title Resultado:
print('DU = %.2f kJ'%(DU/1000))
print('W_DV = %.2f kJ'%(W_DV/1000))
```
