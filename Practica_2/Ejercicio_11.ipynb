# Ejercico 11
Si a través de los límites del sistema contenido en el conjunto cilindro-pistón se entrega un trabajo eléctrico $(W_e)$ de $1000 \, \mathrm{kJ}$. Considerando $\Delta Ec = \Delta Ep = 0$

Determinar:

1. Masa del aire
2. $T_{final}$
3. $W_{ΔV}$

image.png

## Solución

```python
# Datos:
W_e = -1000e3 # J    (trabajo eléctrico)
P = 10e5 # Pa
T_1 = 27 + 273.15 # K
V_1 = 0.3 # m^3
R = 287 # J/(kg*K)  (R del aire, dato de tabla)
cp = 1.0047e3 # J/(kg*K)  (del aire)
Q = 0
```

### 1) Cálculo de masa de aire

Ecuación de estado de los gases ideales
$$ PV = mRT $$

Despejamos $m$
$$ m = \frac{P V_1}{R T_1} $$

```python
# Calculamos la masa con la ecuación de estado de los gases ideales
m = (P*V_1)/(R*T_1)
```

```python
#@title Resultado:
print('m = %.4f kg'%m)
```

### 2) Cálculo de temperatura final

Primer principio de la termodinámica:
$$ Q = \Delta U + W $$

$$ Q = \Delta U + W_{\Delta V} + W_e $$

Tenemos que:
$$ \Delta U + W_{\Delta V} = \Delta H = m \, c_p \Delta T $$

$$ Q = \Delta H + W_e $$

$$ Q = m \, c_p \Delta T + W_e $$

$$ 0 = m \, c_p (T_2 - T_1) + W_e $$

Despejo $T_2$
$$ T_2 = \frac{T_1 \, c_p \, m - W_e}{c_p \, m} $$

```python
T_2 = (T_1*cp*m - W_e)/(cp*m)
```

```python
#@title Resultado:
print('Temperatura final:')
print('T_2 = %.2f K'%T_2)
```

### 3) Cálculo de trabajo por variación de volumen $W_{\Delta V}$

Trabajo por variación de volumen
$$ W_{\Delta V} = P \cdot \Delta V $$

$$ W_{\Delta V} = P \cdot (V_2 - V_1) = P V_2 - P V_1 $$

$$ W_{\Delta V} = m R T_2 - P V_1 $$

Variación de volumen:
$$ \Delta V =\frac{W_{\Delta V}}{P} $$

```python
# Trabajo por variación de volumen
W_DV = m*R*T_2 - P*V_1
# Variación de volumen
DV = W_DV/P
```

```python
DV = W_DV/P
```

```python
#@title Resultado:
print('Trabajo por variación de volumen:')
print('W_DV = %.2f kJ'%(W_DV/1000))
print('Variación de volumen:')
print('DV = %.4f m^3'%DV)
```
